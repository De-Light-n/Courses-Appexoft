### Завдання 5 — Унікальні значення через set

Дано `data = [3, 1, 2, 3, 1, 5, 2]`. Отримайте унікальні значення. Перевірте, чи були дублікати (підказка: порівняйте `len(list)` та `len(set)`). _(Розділ 2.3.)_

```python
data = [3, 1, 2, 3, 1, 5, 2]

unique = set(data)

print("Унікальні значення:", unique)

if len(data) != len(unique):
    print("Дублікати були")
else:
    print("Дублікатів не було")
```

### Завдання 6 — Операції над множинами

Дано `a = {1, 2, 3, 4}` та `b = {3, 4, 5, 6}`. Виведіть їхнє об'єднання, перетин, різницю `a - b` та симетричну різницю. _(Розділ 2.3.)_

```python
# Ваш код тут
a = {1, 2, 3, 4}
b = {3, 4, 5, 6}

print("Об'єднання:", a | b)
print("Перетин:", a & b)
print("Різниця:", a - b)
print("Симетрична різниця:", a ^ b)
```

### Завдання 7 — Словник

Створіть словник про себе з ключами `name`, `age`, `city`. Виведіть значення за ключем `name`. Додайте ключ `email`. Переберіть словник і виведіть усі пари у форматі `ключ -> значення`. _(Розділ 2.4.)_

```python
user = {
    "name": "Bohdan",
    "age": 20,
    "city": "Львів"
}

print(user["name"])

user["email"] = "bogdan.kravets99@gmail.com"

for key, value in user.items():
    print(f"Ключ: {key}, значення: {value}")
```

### Завдання 8 — Частота символів

Порахуйте, скільки разів кожен символ трапляється у рядку `"programming"`, використавши звичайний `dict` (без Counter). Потім перевірте себе через `Counter`. _(Розділ 2.4.)_

```python
# Ваш код тут
from collections import Counter

x = {}
for i in "programming":
    if i in x:
        x[i] += 1
    else:
        x[i] = 1

print(x)
print(Counter("programming")) 
```

### Завдання 9 — Comprehension

Одним list comprehension отримайте квадрати **непарних** чисел від 1 до 20. _(Частина 3.)_

```python
x = [n ** 2 for n in range(1, 21) if n % 2 != 0]

print(x)
```

### Завдання 10 (⭐) — Глибока копія

Дано вкладений список `matrix = [[1, 2], [3, 4]]`. Зробіть його глибоку копію, змініть елемент у копії й доведіть, що оригінал не змінився. _(Частина 4.)_

```python
# Ваш код тут
import copy

matrix = [[1, 2], [3, 4]]

matrix_copy = copy.deepcopy(matrix)

matrix_copy[0][0] = 100

print("Оригінал:", matrix)
print("Копія:", matrix_copy)
```

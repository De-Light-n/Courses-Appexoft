# Частина 1. Умовні конструкції (`if`)

## 1.1. Найпростіший `if`

`if` («якщо») виконує блок коду **тільки тоді**, коли умова істинна (`True`).

Найважливіше в Python: **блоки коду позначаються відступом** (зазвичай **4 пробіли**).
Усе, що з відступом під `if`, належить цьому `if`. Це не просто краса — це синтаксис.

````python
temperature = 30

if temperature > 25:
    print("Спекотно!")          # цей рядок з відступом -> належить if
    print("Увімкни кондиціонер") # теж належить if

print("Кінець програми")         # БЕЗ відступу -> виконається завжди
````

Якщо умова хибна (`False`), блок з відступом просто **пропускається**:

````python
temperature = 15

if temperature > 25:
    print("Спекотно!")     # НЕ виконається, бо 15 > 25 -> False

print("Готово")            # виконається завжди
````

## 1.2. `if` / `else`

`else` («інакше») задає, що робити, коли умова **хибна**. Виконається рівно одна з двох гілок.

````python
age = 16

if age >= 18:
    print("Доступ дозволено")
else:
    print("Доступ заборонено — вам ще немає 18")
````

## 1.3. `if` / `elif` / `else` — кілька варіантів

Коли варіантів більше двох, додають `elif` (скорочення від «else if»). Python перевіряє умови
**по черзі згори вниз** і виконує **першу**, що виявилась істинною. Решту — пропускає.

````python
score = 82

if score >= 90:
    grade = "A"
elif score >= 75:      # перевіряється, лише якщо перша умова була False
    grade = "B"
elif score >= 60:
    grade = "C"
else:                  # якщо жодна умова не підійшла
    grade = "F"

print("Оцінка:", grade)   # B
````

> 🧠 Порядок важливий! Умови перевіряються згори вниз, і виконується **перша** істинна.
> Тому спочатку пишуть «вужчі» умови (більші пороги), а потім «ширші».

## 1.4. Складніші умови

Умови можна поєднувати логічними операторами `and`, `or`, `not` (з Лекції 1),
а також вкладати один `if` в інший.

````python
age = 25
has_ticket = True

# Поєднання умов через and:
if age >= 18 and has_ticket:
    print("Проходьте")
else:
    print("Вхід неможливий")

# Вкладені умови — коли рішення багатоступеневе:
temperature = 30
if temperature > 0:
    if temperature > 25:
        print("Спекотно")
    else:
        print("Тепло")
else:
    print("Мороз")
````

Умови зручно поєднувати з операторами `in` (належність) та ланцюговим порівнянням:

````python
day = "Sat"
if day in ("Sat", "Sun"):        # чи є day серед вихідних?
    print("Вихідний")

x = 50
if 0 <= x <= 100:                # ланцюг: те саме, що (0 <= x) and (x <= 100)
    print("x у діапазоні від 0 до 100")
````

## 1.5. Тернарний оператор (умова в один рядок)

Коли треба обрати одне з двох значень, зручний короткий запис: `X if умова else Y`.
Читається як «`X`, якщо умова, інакше `Y`».

````python
age = 20

# Довгий спосіб:
if age >= 18:
    status = "дорослий"
else:
    status = "неповнолітній"
print(status)

# Те саме одним рядком (тернарний оператор):
status = "дорослий" if age >= 18 else "неповнолітній"
print(status)
````

## 1.6. 🔎 Важливо знати: `match` / `case` (сучасний Python 3.10+)

Коли треба порівняти значення з багатьма варіантами, у Python 3.10+ є конструкція `match`/`case`
(аналог `switch` в інших мовах). Про неї інколи питають на співбесідах.

```python
def http_status(code):
    match code:
        case 200:
            return "OK"
        case 400 | 401 | 403:      # кілька значень через |
            return "Помилка клієнта"
        case 500:
            return "Помилка сервера"
        case _:                    # _ означає "усе інше" (за замовчуванням)
            return "Невідомо"

print(http_status(404))            # Невідомо
```

> ⚠️ `match`/`case` працює лише в **Python 3.10 і новіших**. Те саме завжди можна записати через
> `if`/`elif`/`else`. Приклад тут — у вигляді тексту, щоб зошит не зламався на старіших версіях.

# Частина 2. Цикли

**Цикл** повторює дію багато разів, щоб не писати той самий код вручну. У Python два цикли:
- **`for`** — пройти по всіх елементах чогось (список, рядок, діапазон чисел);
- **`while`** — повторювати, **поки** умова істинна.

## 2.1. Цикл `for`

`for` бере елементи **по черзі** й виконує тіло циклу (блок з відступом) для кожного.

> Тут ми перебираємо **список** (`[...]`) і **рядок**. Повноцінно колекції — у Модулі 2,
> зараз досить знати: список — це набір значень у квадратних дужках.

````python
# Перебір списку:
for fruit in ["яблуко", "груша", "слива"]:
    print("Фрукт:", fruit)

print("---")

# Рядок теж можна перебирати — по символах:
for letter in "Код":
    print(letter)
````

### `range()` — послідовність чисел

Щоб повторити дію задану кількість разів або пройтись по числах, використовують `range()`:
- `range(n)` — числа `0, 1, ..., n-1` (рівно `n` штук);
- `range(start, stop)` — від `start` до `stop` (**`stop` не включається**);
- `range(start, stop, step)` — з кроком.

````python
# range(n): n повторень
for i in range(3):
    print("Повторення номер", i)   # 0, 1, 2

print("---")

# range(start, stop): stop НЕ входить
for i in range(1, 6):
    print(i)                        # 1, 2, 3, 4, 5

print("---")

# range з кроком:
for i in range(0, 11, 2):
    print(i)                        # 0, 2, 4, 6, 8, 10
````

Класичний приклад — накопичити суму в циклі:

````python
total = 0
for number in [10, 20, 30, 40]:
    total = total + number     # додаємо кожне число до суми
    print("Проміжна сума:", total)

print("Разом:", total)         # 100
````

## 2.2. Цикл `while`

`while` повторює тіло, **поки умова істинна**. Використовують, коли **не знаємо заздалегідь**,
скільки разів треба повторити.

````python
# Рахуємо від 5 до 1:
n = 5
while n > 0:
    print(n)
    n = n - 1      # ВАЖЛИВО змінювати n, інакше умова завжди True

print("Старт!")
````

> ⚠️ **Найчастіша помилка — нескінченний цикл.** Якщо умова `while` ніколи не стає `False`,
> програма «зависне» назавжди. Завжди перевіряйте, що всередині циклу є щось, що врешті
> зробить умову хибною (наприклад, `n = n - 1`). Якщо цикл завис — зупиніть його
> (у Jupyter: кнопка ⏹ «Stop» / «Interrupt»).

## 2.3. Керування циклом: `break` і `continue`

- **`break`** — **негайно вийти** з циклу повністю.
- **`continue`** — **пропустити** решту тіла й перейти до наступного елемента.

````python
# break: зупиняємось, щойно знайшли потрібне
for number in [3, 7, 10, 15, 20]:
    if number > 10:
        print("Знайшли перше число більше за 10:", number)
        break                  # виходимо з циклу
    print("перевіряю:", number)
````

````python
# continue: пропускаємо парні числа
for number in range(1, 8):
    if number % 2 == 0:
        continue               # пропустити решту тіла для парних
    print("непарне:", number)  # 1, 3, 5, 7
````

## 2.4. `enumerate` — індекс і елемент разом

Часто потрібен і **номер** елемента, і **сам** елемент. `enumerate` дає обидва одразу —
не треба вручну рахувати індекс.

````python
shopping = ["хліб", "молоко", "яйця"]

# enumerate дає пари (індекс, елемент). start=1 -> нумерація з 1.
for index, item in enumerate(shopping, start=1):
    print(f"{index}) {item}")
````

> 🔎 **Важливо знати (питають на співбесіді):** не пишіть `for i in range(len(список))`,
> а потім `список[i]`. Ідіоматичний Python — це `enumerate`: коротше, читабельніше й надійніше.

## 2.5. `zip` — паралельний перебір кількох списків

`zip` «зшиває» кілька списків і дає їхні елементи **парами** — зручно перебирати два списки
одночасно.

````python
names = ["Аня", "Богдан", "Марія"]
ages = [25, 30, 22]

# Беремо по одному елементу з КОЖНОГО списку одночасно:
for name, age in zip(names, ages):
    print(f"{name} — {age} р.")

# zip зупиняється по НАЙКОРОТШОМУ списку:
print(list(zip([1, 2, 3], ["a", "b"])))   # [(1, 'a'), (2, 'b')]
````

## 2.6. 🔎 Важливо знати: `for` / `else`

У Python цикл `for` може мати блок `else`. Правило незвичне, але його люблять на співбесідах:
**`else` виконується, лише якщо цикл завершився БЕЗ `break`.**

Типове застосування — пошук: `else` = «перебрали все й не знайшли».

````python
numbers = [4, 6, 8, 10]

for n in numbers:
    if n % 2 != 0:                 # шукаємо непарне
        print("Знайдено непарне:", n)
        break
else:
    # виконається, ТІЛЬКИ якщо break жодного разу не спрацював
    print("Непарних чисел немає — усі парні")
````

- `for`/`else` — `else` спрацьовує, якщо не було `break`.

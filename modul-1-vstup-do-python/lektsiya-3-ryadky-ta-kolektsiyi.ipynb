# Частина 1. Рядки (`str`)

## 1.1. Що таке рядок

**Рядок (`str`)** — це **текст**: послідовність символів у лапках. Ім'я, повідомлення, адреса —
усе це рядки. Окремого типу «символ» у Python немає: один символ — це теж рядок завдовжки 1.

````python
city = "Львів"
print(city)
print(type(city))     # <class 'str'>
print(len(city))      # 5 — кількість символів
````

### Способи створити рядок
- в **подвійних** лапках: `"текст"`
- в **одинарних** лапках: `'текст'` (різниці немає, головне — однакові з обох боків)
- **багаторядковий** — у потрійних лапках `"""..."""`

````python
a = "подвійні лапки"
b = 'одинарні лапки'
print(a)
print(b)

# Багаторядковий текст — потрійні лапки:
poem = """Рядок перший
Рядок другий
Рядок третій"""
print(poem)

# Одинарні лапки зручні, коли всередині тексту є подвійні (і навпаки):
quote = 'Він сказав: "Привіт!"'
print(quote)
````

## 1.2. Індексація: доступ до окремих символів

Кожен символ рядка має **номер (індекс)**. Нумерація починається з **0**, а не з 1!
Від'ємні індекси рахують з кінця: `-1` — останній символ.

```
 рядок:   P   y   t   h   o   n
 індекс:  0   1   2   3   4   5
 з кінця:-6  -5  -4  -3  -2  -1
```

````python
word = "Python"
print(word[0])    # 'P' — перший символ (індекс 0)
print(word[1])    # 'y'
print(word[-1])   # 'n' — останній символ
print(word[-2])   # 'o' — передостанній

# len() дає кількість символів; останній індекс = len - 1
print("Довжина:", len(word))   # 6
print("Останній:", word[len(word) - 1])   # 'n'
````

## 1.3. Зрізи (slicing) — беремо частину рядка

Зріз дозволяє взяти **підрядок**. Синтаксис: `рядок[start:stop:step]`
- `start` — з якого індексу почати (**включно**);
- `stop` — до якого дійти (**НЕ включно**);
- `step` — крок (необов'язково).

Будь-яку частину можна пропустити. Це працює не лише з рядками, а й зі списками та кортежами.

````python
word = "Python"
print(word[0:3])    # 'Pyt'  -> символи 0, 1, 2 (3 НЕ входить)
print(word[2:])     # 'thon' -> від індексу 2 і до кінця
print(word[:2])     # 'Py'   -> від початку до індексу 2 (не включно)
print(word[:])      # 'Python' -> копія всього рядка

# Крок (step):
print(word[::2])    # 'Pto'  -> кожен другий символ
print(word[::-1])   # 'nohtyP' -> крок -1 РОЗВЕРТАЄ рядок задом наперед
````

> 💡 Прийом `рядок[::-1]` — найпростіший спосіб **перевернути** рядок. Наприклад, так легко
> перевірити паліндром (слово, що читається однаково в обидва боки):

````python
s = "радар"
print(s[::-1])          # 'радар'
print(s == s[::-1])     # True — отже, це паліндром
````

## 1.4. Корисні методи рядків

**Метод** — це дія, яку можна виконати над значенням через крапку: `значення.метод()`.
У рядків їх багато; ось найпотрібніші. Кожен метод **повертає новий рядок** (оригінал не змінюється).

````python
text = "  Python — Чудова Мова  "

print(text.strip())              # прибрати пробіли по краях
print(text.lower())              # усі літери маленькі
print(text.upper())              # усі літери великі
print(text.strip().title())      # Кожне Слово З Великої

print("---")
print(text.replace("Чудова", "Потужна"))   # замінити підрядок
print("яблуко,груша,слива".split(","))      # розбити рядок -> список
print("-".join(["2024", "07", "15"]))       # зібрати список -> рядок: 2024-07-15
````

````python
name = "Python"
print(name.startswith("Py"))   # True  — починається з "Py"?
print(name.endswith("on"))     # True  — закінчується на "on"?
print(name.find("th"))         # 2     — індекс, де починається "th" (або -1, якщо нема)
print("banana".count("a"))     # 3     — скільки разів трапляється "a"

# Перевірки вмісту:
print("12345".isdigit())       # True  — рядок складається лише з цифр?
print("Hello".isalpha())       # True  — лише літери?
````

## 1.5. Форматування тексту: f-strings

Найзручніший спосіб вставити значення змінних у текст — **f-рядок**. Ставимо букву `f` перед
лапками й пишемо змінні або вирази у фігурних дужках `{}`.

````python
name = "Ноутбук"
price = 24999.5

# Просто вставляємо значення:
print(f"Товар: {name}, ціна: {price} грн")

# Усередині {} можна писати вирази:
print(f"Дві ціни: {price * 2}")

# Форматування чисел після двокрапки:
print(f"{price:.2f}")     # 24999.50  -> 2 знаки після коми
print(f"{price:,.2f}")    # 24,999.50 -> роздільник тисяч
print(f"{0.1234:.1%}")    # 12.3%     -> у відсотках
````

## 1.6. 🔎 Важливо знати: рядок **не можна змінити** (immutable)

Рядок в Python **незмінний** (*immutable*). Це означає: після створення його вміст
змінити **не можна** — можна лише створити **новий** рядок. Спроба змінити символ дасть помилку:

> ℹ️ У прикладі нижче використано `try`/`except` — це «перехоплення» помилки, щоб зошит не
> зупинявся. Повноцінно обробку помилок розберемо **далі в курсі**; тут просто ловимо помилку, щоб її показати.

````python
s = "abc"
try:
    s[0] = "X"        # спроба змінити перший символ
except TypeError as e:
    print("Помилка:", e)   # 'str' object does not support item assignment
````

А коли ми «додаємо» до рядка — насправді створюється **новий** об'єкт у пам'яті.
Побачити це можна через `id()` (умовно — «адреса» об'єкта в пам'яті): вона **змінюється**.

````python
s = "abc"
print("id до :", id(s))
s = s + "d"                 # це НЕ зміна, а створення нового рядка "abcd"
print("id після:", id(s))   # інша адреса -> це вже інший об'єкт
````

**Чому це важливо (і що можуть спитати на співбесіді):**
- Оскільки рядок незмінний, його можна безпечно використовувати як **ключ словника** (див. далі).
- Якщо в циклі склеювати тисячі рядків через `+`, щоразу створюється новий об'єкт — це **повільно**.
  Правильно — накопичити частини у список і один раз з'єднати через `"".join(...)`.

📎 Наочно «побачити» об'єкти в пам'яті можна тут: <https://pythontutor.com/>

## 1.7. 🔎 Важливо знати: `str` vs `bytes` та Unicode

Ще одна тема, яку люблять на співбесідах.
- **`str`** — це **текст** (символи, які бачить людина).
- **`bytes`** — це **сирі байти** (як дані зберігаються у файлі чи передаються мережею).
- `str.encode()` перетворює текст → байти, `bytes.decode()` → назад. Стандартне кодування — **UTF-8**.

Важливо: в UTF-8 українська літера займає **2 байти**, тому «довжина в символах» ≠ «довжина в байтах».

````python
s = "Привіт"
b = s.encode("utf-8")          # текст -> байти
print("Байти:", b)             # b'\xd0\x9f...' — так виглядають сирі дані
print("Символів:", len(s))     # 6  — стільки бачить людина
print("Байтів: ", len(b))      # 12 — кирилиця = 2 байти на символ
print("Назад у текст:", b.decode("utf-8"))
````

# Частина 2. Колекції

**Колекція** — це «контейнер», що зберігає багато значень в одній змінній. У Python
чотири основні колекції. Ось коротке порівняння — а далі розберемо кожну окремо й детально.

| Тип | Приклад | Впорядкований? | Можна змінювати? | Дублікати? | Для чого |
|---|---|---|---|---|---|
| **`list`** (список) | `[1, 2, 3]` | так | так | так | послідовність, що змінюється |
| **`tuple`** (кортеж) | `(1, 2, 3)` | так | **ні** | так | незмінний набір значень |
| **`set`** (множина) | `{1, 2, 3}` | ні | так | **ні** | унікальні значення, швидкий пошук |
| **`dict`** (словник) | `{"a": 1}` | так | так | ключі унікальні | пари «ключ → значення» |

> «Впорядкований» тут означає: чи зберігається **порядок**, у якому ви додавали елементи.

## 🔁 Маленький відступ: цикл `for`

Щоб зробити щось **з кожним елементом** колекції, використовують **цикл `for`** — він проходить
по елементах по черзі. Ви вже вивчали цикли в **Модулі 1 (Лекція 2)** — тут коротке нагадування,
бо далі вони нам знадобляться постійно.

````python
# for бере елементи ПО ЧЕРЗІ й виконує відступлений код для кожного:
for fruit in ["яблуко", "груша", "слива"]:
    print("Фрукт:", fruit)

print("---")

# range(n) генерує числа 0, 1, ..., n-1 — зручно, щоб повторити дію n разів:
for i in range(3):
    print("Крок номер", i)
````

## 2.1. `list` — список

**Список** — найуніверсальніша колекція: впорядкований набір значень, який можна змінювати.
Створюється квадратними дужками `[]`. Елементи можуть бути будь-яких типів.

````python
fruits = ["яблуко", "груша", "слива"]
print(fruits)
print("Кількість:", len(fruits))

# Доступ за індексом і зрізи — так само, як у рядків:
print(fruits[0])     # 'яблуко' — перший
print(fruits[-1])    # 'слива'  — останній
print(fruits[0:2])   # ['яблуко', 'груша'] — зріз

# Список МОЖНА змінювати (на відміну від рядка):
fruits[0] = "банан"
print(fruits)        # ['банан', 'груша', 'слива']
````

### Основні методи списку

````python
nums = [3, 1, 2]

nums.append(4)        # додати В КІНЕЦЬ -> [3, 1, 2, 4]
nums.insert(0, 0)     # вставити за індексом -> [0, 3, 1, 2, 4]
nums.remove(1)        # видалити перше значення 1 -> [0, 3, 2, 4]
last = nums.pop()     # видалити й повернути останній -> last=4
print("Список:", nums, "| знято:", last)

# Сортування:
nums.sort()                     # сортує САМ СПИСОК на місці
print("Після sort():", nums)
print("sorted():", sorted([5, 3, 9], reverse=True))  # повертає НОВИЙ список

# Корисні функції:
print("сума:", sum([1, 2, 3]), "| макс:", max([1, 2, 3]), "| довжина:", len([1, 2, 3]))
````

> ⚠️ **Уважно:** `nums.sort()` сортує список **на місці** й повертає `None`, а не новий список.
> Тому `x = [3,1,2].sort()` дасть `x = None`. Якщо потрібен новий відсортований список —
> беріть `sorted(...)`.

````python
result = [3, 1, 2].sort()
print("sort() повернув:", result)     # None!
print("sorted() повернув:", sorted([3, 1, 2]))   # [1, 2, 3]
````

### 🔎 Важливо знати: як список влаштований усередині

Усередині `list` — це **динамічний масив** (*dynamic array*): суцільний блок пам'яті з посиланнями
на елементи, що вміє автоматично збільшуватись, коли ви додаєте нові елементи.

Що з цього випливає (і про що питають на співбесідах):
- **Доступ за індексом `lst[i]` — миттєвий, O(1)**: Python одразу знає, де лежить елемент.
- **`append()` у кінець — швидкий** (у середньому O(1)).
- **`insert(0, x)` чи `remove(x)` на початку — повільні, O(n)**: усі елементи треба зсунути.
- **Перевірка `x in lst` — повільна, O(n)**: Python перебирає елементи по черзі. Якщо часто
  перевіряєте наявність — краще `set` (див. далі).

> O(1) означає «час не залежить від розміру», O(n) — «час зростає пропорційно кількості елементів».

## 2.2. `tuple` — кортеж

**Кортеж** — це майже те саме, що список, але **незмінний**: після створення його не можна
змінити. Створюється круглими дужками `()`. Використовують для наборів, які не мають мінятися
(координати, дата), і для повернення кількох значень із функції.

````python
point = (10, 20)
print(point)
print(point[0], point[1])   # доступ за індексом — як у списку

# Але змінити НЕ можна:
try:
    point[0] = 99
except TypeError as e:
    print("Помилка:", e)    # 'tuple' object does not support item assignment
````

### Розпакування (unpacking) — дуже зручний прийом
Можна одразу «розкласти» кортеж (чи список) по окремих змінних:

````python
point = (10, 20)
x, y = point            # x отримає 10, y отримає 20
print("x =", x, "| y =", y)

# Обмін значень місцями — в один рядок:
a, b = 1, 2
a, b = b, a
print("a =", a, "| b =", b)   # a=2, b=1

# Зірочка збирає "решту" в список:
first, *middle, last = [1, 2, 3, 4, 5]
print(first, middle, last)    # 1 [2, 3, 4] 5
````

> ⚠️ **Пастка:** кортеж з **одного** елемента вимагає коми! `(5)` — це просто число `5`,
> а `(5,)` — кортеж. Саме кома робить кортеж, а не дужки.

````python
not_a_tuple = (5)     # це int
real_tuple = (5,)     # це кортеж з одного елемента
print(type(not_a_tuple))   # <class 'int'>
print(type(real_tuple))    # <class 'tuple'>
````

### 🔎 Важливо знати: навіщо взагалі кортеж, якщо є список?
- **Безпека:** незмінність гарантує, що дані випадково не зіпсують.
- **Швидкість:** кортеж трохи легший і швидший за список.
- **Може бути ключем словника / елементом множини** — а список не може (бо змінний). Чому саме —
  стане зрозуміло в розділі про `dict`.

## 2.3. `set` — множина

**Множина** — колекція **унікальних** елементів **без порядку**. Створюється фігурними
дужками `{}`. Два головні застосування: **позбутися дублікатів** і **дуже швидко перевіряти**,
чи є елемент у наборі.

````python
# Дублікати автоматично зникають:
nums = [1, 2, 2, 3, 3, 3]
unique = set(nums)
print(unique)          # {1, 2, 3}

# Порядок не зберігається — множина невпорядкована:
letters = {"c", "a", "b"}
print(letters)         # може вивестись у будь-якому порядку

# Перевірка наявності — дуже швидка:
print(2 in unique)     # True
print(9 in unique)     # False
````

### Операції над множинами (як у математиці)

Уявіть два кола, що частково перетинаються — це і є множини `A` і `B`:

```
        A = {1, 2, 3}                 B = {3, 4, 5}

           ┌──────────┐
       ┌───┼──────┐   │
       │ 1 │  3   │ 4 │
       │ 2 │      │ 5 │
       └───┼──────┘   │
           └──────────┘
       тільки  спільне  тільки
        в A     (3)      в B

  A | B  (об'єднання)        = {1, 2, 3, 4, 5}   усе разом
  A & B  (перетин)           = {3}               лише спільне
  A - B  (різниця)           = {1, 2}            є в A, немає в B
  A ^ B  (симетрична різниця)= {1, 2, 4, 5}      усе, КРІМ спільного
```

````python
A = {1, 2, 3}
B = {3, 4, 5}

print("об'єднання  A | B:", A | B)   # {1, 2, 3, 4, 5}
print("перетин     A & B:", A & B)   # {3}
print("різниця     A - B:", A - B)   # {1, 2}
print("симетрична  A ^ B:", A ^ B)   # {1, 2, 4, 5}
````

### 🔎 Важливо знати: як множина працює і чому пошук такий швидкий

`set` усередині — це **хеш-таблиця** (та сама технологія, що й у словника, див. наступний розділ).
Замість того щоб перебирати елементи по черзі, Python обчислює для елемента спеціальне число —
**хеш** — і за ним одразу знаходить потрібну «комірку».

Наслідки (важливо для співбесід):
- **`x in set` — у середньому O(1)** (миттєво), тоді як `x in list` — O(n) (перебір).
  Тому для частих перевірок «чи є елемент» обирайте `set`.
- У множину можна класти **лише незмінні (hashable)** значення: числа, рядки, кортежі.
  **Список покласти не можна** — буде помилка:

````python
# Порівняння швидкості ідеї: set знаходить елемент миттєво, list — перебором.
print(1000 in {1000})     # True (O(1))

# У множину не можна покласти список (він змінний -> не хешований):
try:
    bad = {[1, 2]}
except TypeError as e:
    print("Помилка:", e)   # unhashable type: 'list'
````

## 2.4. `dict` — словник

**Словник** зберігає пари **«ключ → значення»** і дозволяє миттєво отримати значення за ключем.
Це, мабуть, найважливіша структура даних у Python. Створюється фігурними дужками з двокрапками:
`{ключ: значення}`.

````python
user = {"name": "Ірина", "age": 25}

# Доступ до значення за ключем:
print(user["name"])              # Ірина

# Безпечний доступ через .get() — не впаде, якщо ключа немає:
print(user.get("email", "—"))    # — (значення за замовчуванням)

# Додати / оновити:
user["email"] = "iryna@mail.com" # додали новий ключ
user["age"] = 26                 # оновили наявний
print(user)

# Видалити ключ:
del user["age"]
print(user)
````

### Перебір словника

````python
user = {"name": "Ірина", "age": 25, "city": "Львів"}

# По парах ключ-значення (найчастіше так):
for key, value in user.items():
    print(f"  {key} -> {value}")

print("Ключі:", list(user.keys()))       # ['name', 'age', 'city']
print("Значення:", list(user.values()))  # ['Ірина', 25, 'Львів']
````

### 🔎 Важливо знати: як словник влаштований усередині (хеш-таблиця)

Це **найпопулярніше питання про Python на співбесідах**, тож розберемо наочно.

Коли ви пишете `d["name"] = "Іра"`, відбувається таке:

```
   ключ "name"
       │
       ▼  hash("name")  -> велике число, напр. 4823...
       │
       ▼  число вказує на "комірку" у внутрішньому масиві
   ┌─────┬─────┬─────┬───────────────────┬─────┐
   │  0  │  1  │  2  │ 3: ("name","Іра") │  4  │  ...
   └─────┴─────┴─────┴───────────────────┴─────┘

   Пошук d["name"]:  знову рахуємо hash("name") -> одразу комірка №3
                     -> беремо значення. Не треба перебирати всі ключі!
```

1. Python рахує **`hash(ключ)`** — велике число.
2. За цим числом визначає **комірку** в масиві, куди покласти пару.
3. При пошуку `d[ключ]` рахує той самий хеш → одразу йде в потрібну комірку → **O(1)**.
4. Якщо два різні ключі дали ту саму комірку — це **колізія** (*collision*). Тоді Python
   порівнює ключі через `==` і шукає інше вільне місце. CPython використовує стратегію
   **відкритої адресації** (*open addressing*): пробує наступні комірки за спеціальною
   послідовністю, поки не знайде вільну. (Інші мови часто застосовують **ланцюжки** —
   *chaining* — список елементів у кожній комірці.)

**Головні наслідки (запам'ятайте для співбесіди):**
- Доступ за ключем — у середньому **O(1)** (миттєвий), як і `in` для `dict`/`set`.
- У **найгіршому** випадку (багато колізій) швидкість може впасти до **O(n)**, але на практиці — O(1).
- **Ключем може бути лише незмінне (hashable) значення**: `str`, `int`, `float`, `bool`, `tuple`.
  **Список чи інший словник ключем бути не можуть** — вони змінні, їхній хеш «поплив би».
- Правило: якщо `a == b`, то обов'язково `hash(a) == hash(b)` (інакше словник «загубив» би ключ).

📎 Наочно про хеш-таблиці й колізії: <https://www.geeksforgeeks.org/hashing-data-structure/>
📎 Візуалізація хеш-таблиці (інтерактивно): <https://www.cs.usfca.edu/~galles/visualization/OpenHash.html>

````python
# hash() працює для незмінних значень:
print("hash('abc'):", hash("abc"))
print("hash((1, 2)):", hash((1, 2)))

# Кортеж можна за ключ (незмінний), а список — ні:
coords = {(50.45, 30.52): "Київ", (49.84, 24.03): "Львів"}
print(coords[(49.84, 24.03)])   # Львів

try:
    bad = {[1, 2]: "значення"}   # список ключем -> помилка
except TypeError as e:
    print("Помилка:", e)         # unhashable type: 'list'
````

### 🔎 Цікава деталь для співбесіди: `True`, `1` і `1.0` — «один ключ»

Оскільки `hash(True) == hash(1) == hash(1.0)` **і** `True == 1 == 1.0`, словник вважає їх
**одним і тим самим ключем**. Тому три присвоєння нижче «зіллються» в один запис:

````python
d = {}
d[1] = "ціле"
d[True] = "булеве"
d[1.0] = "дробове"
print(d)        # {1: 'дробове'} — усі три "злилися", лишилось останнє значення
````

### Порядок вставки та зручні інструменти

Починаючи з **Python 3.7**, словник **гарантовано зберігає порядок**, у якому ви додавали ключі
(це не сортування — саме порядок додавання). А модуль `collections` дає корисні різновиди словників:

````python
# import підключає готовий модуль (див. Лекцію 1). collections — модуль зі стандартної бібліотеки.
from collections import Counter, defaultdict

# Порядок вставки зберігається:
d = {}
for ch in "bca":
    d[ch] = ord(ch)    # ord(символ) -> його номер у таблиці Unicode (напр. 'a' -> 97)
print("Порядок:", list(d))    # ['b', 'c', 'a'] — як додавали

# Counter — миттєво порахувати, скільки разів трапляється кожен елемент:
print(Counter("mississippi"))     # {'i': 4, 's': 4, 'p': 2, 'm': 1}

# defaultdict — словник, що сам створює значення за замовчуванням (тут — порожній список).
# Зручно для групування:
groups = defaultdict(list)
for word in ["apple", "avocado", "banana", "cherry"]:
    groups[word[0]].append(word)   # групуємо слова за першою літерою
print(dict(groups))
````

# 🔎 Важливо знати: швидкодія колекцій (Big-O)

Ми вже кілька разів згадували `O(1)` та `O(n)`. Це **Big-O** — спосіб описати, як швидкодія
операції залежить від кількості елементів `n`. Це **обов'язкова тема для співбесід**.

Простими словами:
- **O(1)** — «сталий час»: **не залежить** від розміру. Наприклад, `d[key]` у словнику —
  однаково швидко і для 10, і для 10 000 000 елементів.
- **O(n)** — «лінійний»: час зростає **пропорційно** кількості елементів. Наприклад, `x in list` —
  Python перебирає елементи по черзі; удвічі більший список — удвічі довше.
- **O(log n)** — «логарифмічний»: дуже повільно зростає (двійковий пошук, збалансовані дерева).
- **O(n²)** — «квадратичний»: вкладений цикл по тих самих даних; на великих `n` стає болючим.

### Шпаргалка: складність типових операцій (у середньому)

| Операція | `list` | `tuple` | `set` | `dict` |
|---|---|---|---|---|
| доступ за індексом `x[i]` | **O(1)** | **O(1)** | — | — |
| доступ за ключем `x[k]` | — | — | — | **O(1)** |
| пошук `елемент in x` | **O(n)** | **O(n)** | **O(1)** | **O(1)** (по ключу) |
| додати в кінець `append` | **O(1)** | незмінний | `add` **O(1)** | `x[k]=v` **O(1)** |
| вставити/видалити на початку | **O(n)** | незмінний | — | — |
| довжина `len(x)` | **O(1)** | **O(1)** | **O(1)** | **O(1)** |

**Практичний висновок:** якщо ви **часто перевіряєте наявність** елемента — беріть `set`/`dict`
(**O(1)**), а не `list` (**O(n)**). Це одна з найтиповіших оптимізацій, яку питають на співбесідах.

````python
# Наочно: чому "in" для set набагато вигідніше за list на великих даних.
# (Обидва дадуть True, але set шукає миттєво, а list — перебором.)
big_list = list(range(1_000_000))
big_set = set(big_list)

print(999_999 in big_list)   # O(n): Python пройде майже весь список
print(999_999 in big_set)    # O(1): миттєво через хеш
````

> ⏱️ Хочете **побачити** різницю в часі? У Jupyter є «магічна» команда `%timeit`, що вимірює
> швидкість. Запустіть комірку нижче — час для `set` буде в тисячі разів меншим.

````python
big_list = list(range(1_000_000))
big_set = set(big_list)

# %timeit — інструмент Jupyter для заміру часу виконання рядка:
%timeit 999_999 in big_list
%timeit 999_999 in big_set
````

# Частина 3. Comprehensions — компактне створення колекцій

**Comprehension** (включення) — короткий спосіб створити колекцію на основі іншої.
Читається як «**візьми X для кожного елемента з Y**». Спершу порівняймо зі звичайним циклом:

````python
# Звичайний спосіб — через цикл:
squares = []
for n in range(1, 6):
    squares.append(n ** 2)
print(squares)     # [1, 4, 9, 16, 25]

# Те саме одним рядком через list comprehension:
squares = [n ** 2 for n in range(1, 6)]
print(squares)     # [1, 4, 9, 16, 25]
````

Можна додати **умову** (фільтр) — залишити тільки потрібні елементи. І так само працює
для словників і множин:

````python
# З умовою: тільки парні числа
evens = [n for n in range(10) if n % 2 == 0]
print(evens)       # [0, 2, 4, 6, 8]

# dict comprehension — порахувати ціни з ПДВ:
prices = {"кава": 45, "чай": 30}
with_vat = {name: round(p * 1.2, 2) for name, p in prices.items()}   # round(x, 2) -> округлити до 2 знаків
print(with_vat)    # {'кава': 54.0, 'чай': 36.0}

# set comprehension — унікальні довжини слів:
lengths = {len(w) for w in ["a", "bb", "cc", "ddd"]}
print(lengths)     # {1, 2, 3}
````

# Частина 4. Копіювання: поверхневе vs глибоке 🔎

Дуже важлива тема (і часте питання на співбесідах). Згадаймо з Лекції 1: `b = a` — це **не копія**,
а **другий ярлик** на той самий об'єкт. Зміна через одне ім'я видно і через друге:

````python
a = [1, 2, 3]
b = a              # НЕ копія — просто друге ім'я того самого списку
b.append(4)
print("a:", a)     # [1, 2, 3, 4] — змінився теж!
print("a is b:", a is b)   # True — це один об'єкт
````

Щоб отримати **справжню копію**, є два способи:
- **Поверхнева копія** (`.copy()`, `list(a)`, `a[:]`) — новий зовнішній список, але **вкладені**
  об'єкти лишаються **спільними**.
- **Глибока копія** (`copy.deepcopy(a)`) — копіює **все** вкладене рекурсивно, повна незалежність.

````python
import copy

# Поверхнева копія "протікає" на вкладених списках:
original = [[1, 2], [3, 4]]
shallow = original.copy()
shallow[0].append(99)              # міняємо ВКЛАДЕНИЙ список
print("shallow -> original:", original)   # [[1, 2, 99], [3, 4]] — зачепило оригінал!

# Глибока копія повністю незалежна:
original = [[1, 2], [3, 4]]
deep = copy.deepcopy(original)
deep[0].append(99)
print("deep    -> original:", original)   # [[1, 2], [3, 4]] — оригінал цілий
````

**Висновок:** якщо структура **вкладена** (список списків, список словників) і потрібна
повна незалежність копії — використовуйте `copy.deepcopy()`. Для простих «плоских» списків
достатньо `.copy()`.

**Колекції — коротко коли що:**
- **`list`** `[...]` — впорядкований, змінний; універсальний. Усередині — динамічний масив.

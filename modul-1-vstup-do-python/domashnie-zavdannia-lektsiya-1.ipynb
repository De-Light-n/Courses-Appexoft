### Завдання 2 — Калькулятор

Створіть дві змінні з числами `a = 15` та `b = 4`. Виведіть їхню суму, різницю, добуток, результат ділення, цілочисельного ділення `//` та остачу `%`. Підпишіть кожен результат. _(Розділ 9.1.)_

```python
a, b = 15, 4
print("Сума:", a + b)
print("Різниця:", a - b)
print("Добуток:", a * b)
print("Ділення:", a / b)
print("Цілочисельне ділення:", a // b)
print("Остача від ділення:", a % b)
```

### Завдання 3 — Парне чи непарне

Задайте число у змінній `n`. За допомогою остачі `%` та порівняння `==` виведіть `True`, якщо число парне, і `False`, якщо непарне. _(Розділи 9.1–9.2.)_

```python
n = 13
print(n % 2 == 0)
```

### Завдання 4 — Перетворення типів

Дано рядок `text = "350"`. Перетворіть його на число, додайте `150` і виведіть результат. Потім перетворіть число `42` на рядок і склейте його з текстом `"Відповідь: "`. _(Розділ 7.)_

```python
text = 350

a = int(text)
print(a + 150)

b = str(42)
print('Відповідь: ' + b)
```

### Завдання 6 — Логіка

Задайте змінні `age = 20` та `has_ticket = True`. Виведіть `True`, якщо людині є 18 або більше **і** вона має квиток. Спробуйте також з `or` та `not`. _(Розділ 9.3.)_

```python
age = 20
has_ticket = True

print(age >= 18 and has_ticket)
print(age >= 18 or has_ticket)
print(not has_ticket)
```

### Завдання 7 — Правдивість

Для кожного зі значень `0`, `""`, `"привіт"`, `None`, `42` виведіть результат `bool(...)`. Поясніть коментарем, чому кожне з них `True` чи `False`. _(Розділ 10.)_

```python
print(bool('')) #False - порожнє значення
print(bool(0)) #False 0
print(bool('hello')) #True - непорожній рядок
print(bool(42)) #True не нульове число
```

### Завдання 8 (⭐ необов'язкове) — `==` проти `is`

Створіть два однакові списки `x = [1, 2]` та `y = [1, 2]`. Виведіть результат `x == y` та `x is y`. Поясніть коментарем різницю. _(Розділ 11.)_

```python
x = [1, 2]
y = [1, 2]

print(x == y) #виведе True - тому що подвійне дорівнює порівнює значення змінних, а вони однакові у даному випадку
print(x is y) #виведе False - is перевіряє чи це той самий обєкт у памяті (було б тру, якби одна зі змінних була б рівна іншій (х = у))
```

### Завдання 2 — Калькулятор

Створіть дві змінні з числами `a = 15` та `b = 4`. Виведіть їхню суму, різницю, добуток, результат ділення, цілочисельного ділення `//` та остачу `%`. Підпишіть кожен результат. _(Розділ 9.1.)_

```python
a = 15
b = 4
print ("Сума:", a + b)
print ("Різниця:", a - b)
print ("Добуток:", a * b)
print ("Ділення:", a / b)
print ("Ділення цілочисельне:", a // b)
print ("Ділення з остачею:", b % a)
```

### Завдання 6 — Логіка

Задайте змінні `age = 20` та `has_ticket = True`. Виведіть `True`, якщо людині є 18 або більше **і** вона має квиток. Спробуйте також з `or` та `not`. _(Розділ 9.3.)_

```python
age = 20
has_ticket = True

print("Повнолітній та має кваток:", age >= 18 and has_ticket)
print("Повнолітній або має кваток:", age >= 18 or has_ticket)
print("Повнолітній:", age >= 18, "Без квитка:", not has_ticket)
```

### Завдання 7 — Правдивість

Для кожного зі значень `0`, `""`, `"привіт"`, `None`, `42` виведіть результат `bool(...)`. Поясніть коментарем, чому кожне з них `True` чи `False`. _(Розділ 10.)_

```python
print (bool(0)) # число 0 є хибним значенням бо 0 це нічого
print (bool("")) # "" порожній рядок є хибним значенням бо нічого нема
print (bool("привіт")) # "привіт" є правдивим значенням тому що в лапках міститься текст
print (bool(None)) # None є хибним значенням тому що None це просто порожнеча, нічого, пусто.
print (bool(42)) # 42 є правдивим значенням тому що це число а все що не 0 правдиве.

# Загалом якщо щось є це True якщо нічого нема (None) чи порожньо ("" []) або число 0 це False
```

### Завдання 8 (⭐ необов'язкове) — `==` проти `is`

Створіть два однакові списки `x = [1, 2]` та `y = [1, 2]`. Виведіть результат `x == y` та `x is y`. Поясніть коментарем різницю. _(Розділ 11.)_

```python
x = [1, 2]
y = [1, 2]

print(x == y) # Перевіряє чи однакові значення
print(x is y) # Перевіряє чи однакові об'єкти 
```

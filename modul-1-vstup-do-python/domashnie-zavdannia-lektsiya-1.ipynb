### Завдання 2 — Калькулятор

Створіть дві змінні з числами `a = 15` та `b = 4`. Виведіть їхню суму, різницю, добуток, результат ділення, цілочисельного ділення `//` та остачу `%`. Підпишіть кожен результат. _(Розділ 9.1.)_

### Завдання 6 — Логіка

Задайте змінні `age = 20` та `has_ticket = True`. Виведіть `True`, якщо людині є 18 або більше **і** вона має квиток. Спробуйте також з `or` та `not`. _(Розділ 9.3.)_

```python
age = 20
has_ticket = True

print(age >= 18 and has_ticket)
print(age >= 18 and not has_ticket)
print(age >= 18 or not has_ticket)
print(age >= 18 or has_ticket)
print(age <= 18 or not has_ticket)
```

### Завдання 7 — Правдивість

Для кожного зі значень `0`, `""`, `"привіт"`, `None`, `42` виведіть результат `bool(...)`. Поясніть коментарем, чому кожне з них `True` чи `False`. _(Розділ 10.)_

```python
# Усі числа та не пусті рядки вважаються правдивими (1 - True), крім "0", порожніх рядків та None (0 -False)
print(bool(0)) 
print(bool(""))
print(bool("привіт"))
print(bool(None))
print(bool(42))
```

### Завдання 8 (⭐ необов'язкове) — `==` проти `is`

Створіть два однакові списки `x = [1, 2]` та `y = [1, 2]`. Виведіть результат `x == y` та `x is y`. Поясніть коментарем різницю. _(Розділ 11.)_

```python
# У першому випадку з == ми порівнюємо значення списків які є однаковими. У другому випадку з is ми порівнюємо два різні об'єкти які є схожі за вмістом 
x = [1, 2, 3]
y = [1, 2, 3]

print(x == y)
print(x is y)
```

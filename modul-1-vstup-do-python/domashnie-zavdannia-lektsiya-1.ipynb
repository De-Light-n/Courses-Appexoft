### Завдання 2 — Калькулятор

Створіть дві змінні з числами `a = 15` та `b = 4`. Виведіть їхню суму, різницю, добуток, результат ділення, цілочисельного ділення `//` та остачу `%`. Підпишіть кожен результат. _(Розділ 9.1.)_

```python
a = 15
b = 4
print(f'Сума чисел {a} та {b} дорівнює {a + b}')
print(f'Різниця чисел {a} та {b} дорівнює {a - b}')
print(f'Добуток чисел {a} та {b} дорівнює {a * b}')
print(f'Частка чисел {a} та {b} дорівнює {a / b}')
print(f'Ціле ділення чисел {a} та {b} дорівнює {a // b}')
print(f'Залишок від ділення чисел {a} та {b} дорівнює {a % b}')
```

### Завдання 7 — Правдивість

Для кожного зі значень `0`, `""`, `"привіт"`, `None`, `42` виведіть результат `bool(...)`. Поясніть коментарем, чому кожне з них `True` чи `False`. _(Розділ 10.)_

```python
print(bool(0)) # false - число 0 завжди вважається False
print(bool('')) # false - порожній рядок завжди вважається False
print(bool('привіт')) # true - рядок з текстомзавжди вважається True
print(bool(None)) # false - None завжди вважається False тому що пайтон завжди його таким вважає для нього вона як порожнеча
print(bool(42)) # true - будь-яке число (окрім 0) завжди вважається True
```

### Завдання 8 (⭐ необов'язкове) — `==` проти `is`

Створіть два однакові списки `x = [1, 2]` та `y = [1, 2]`. Виведіть результат `x == y` та `x is y`. Поясніть коментарем різницю. _(Розділ 11.)_

```python
x = [1,2]
y = [1,2]
print(x ==y) # перевірка через == перевіряє значення, тому повертає True а зміст списків однаковий
print(x is y) # перевірка через is перевіряє саме не значення а об'єкт, тому повертає False бо це різні об'єкти в пам'яті
```

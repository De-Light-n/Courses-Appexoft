### Завдання 6 — Логіка

Задайте змінні `age = 20` та `has_ticket = True`. Виведіть `True`, якщо людині є 18 або більше **і** вона має квиток. Спробуйте також з `or` та `not`. _(Розділ 9.3.)_

```python
age = 20
has_ticket = True

res1 = (age >= 18) and has_ticket
res2 = (age >= 18) and (not has_ticket)
res3 = (age >= 18) or (not has_ticket)
res4 = (age >= 18) or has_ticket
res5 = (age <= 18) or (not has_ticket)

print(res1)
print(res2)
print(res3)
print(res4)
print(res5)
```

### Завдання 7 — Правдивість

Для кожного зі значень `0`, `""`, `"привіт"`, `None`, `42` виведіть результат `bool(...)`. Поясніть коментарем, чому кожне з них `True` чи `False`. _(Розділ 10.)_

```python
print(bool(0))          # - бо нуль вважається порожнім
print(bool(""))         # - бо рядок порожній без жодного символу
print(bool("привіт"))   # + бо рядок не порожній
print(bool(None))       # - бо None означає повну відсутність значення
print(bool(42))         # + будь-яке число яке не нуль це правда
```

### Завдання 8 (⭐ необов'язкове) — `==` проти `is`

```python
x = [1, 2]
y = [1, 2]

print(x == y)
print(x is y)
```

### Завдання 6 — Логіка

### Завдання 7 — Правдивість

Для кожного зі значень `0`, `""`, `"привіт"`, `None`, `42` виведіть результат `bool(...)`. Поясніть коментарем, чому кожне з них `True` чи `False`. _(Розділ 10.)_

```python
# Ваш код тут
print(bool(0)) # False тому що 0 вважається False
print(bool(1)) # True тому що 1 це не пустий рядок і не нуль а ціле число яке вважається True
print(bool("")) # False тому що пустий рядок вважається False
print(bool("Hello")) # True тому що рядок не пустий і вважається True
print(bool(None)) # False тому що None вважається False
print(bool(42)) # True тому що 42 це не пустий рядок і не нуль а ціле число яке вважається True
```

### Завдання 8 (⭐ необов'язкове) — `==` проти `is`

Створіть два однакові списки `x = [1, 2]` та `y = [1, 2]`. Виведіть результат `x == y` та `x is y`. Поясніть коментарем різницю. _(Розділ 11.)_

```python
# Ваш код тут
x = [1, 2]
b = x
y = [1, 2]

print(x == y)  # True, тому що обидва списки мають однакові елементи
print(x is y)  # False, тому що x і y є різними об'єктами в пам'яті
print(x is b)  # True, тому що x і b є одним і тим же об'єктом в пам'яті
```

### Завдання 4 — *args

```python
def maximum(*args):
    a = 0
    for c in args:
        if a < c:
            a = c
    return a

b = maximum(1, 6, 11, 16, 21, 26, 31, 36, 41)
print("Maximum in b:", b)
```

### Завдання 7 — Замикання

Напишіть `make_accumulator()`, що повертає функцію: вона додає передане число до внутрішньої суми й повертає поточну суму. _(Частина 4.)_

```python
def make_accumulator():
    total = 0
    def accumulator(number):
        nonlocal total
        total += number
        return total
    return accumulator

acc = make_accumulator()
print(acc(5))
print(acc(10))
```

### Завдання 8 — Декоратор @timeit

Напишіть декоратор `@timeit`, який друкує час виконання обгорнутої функції (використайте `time.perf_counter`). Не забудьте `functools.wraps`. _(Частина 5.)_

```python
import time
from functools import wraps

def timeit(func):
    @wraps(func)
    def wrapper(*args, **kwargs):
        start = time.perf_counter()
        result = func(*args, **kwargs)
        end = time.perf_counter()
        print(f"Execution time: {end - start} seconds")
        return result
    return wrapper

@timeit
def test_func():
    time.sleep(1)

test_func()
```

### Завдання 9 — Генератор

Напишіть генератор `even_numbers(limit)`, який ліниво видає парні числа до `limit`. Доведіть через `sys.getsizeof`, що він компактніший за відповідний список. _(Частина 6.)_

```python
import sys


def even_numbers(limit):
    for i in range(0, limit + 1, 2):
        yield i


gen = even_numbers(10000)
lst = list(range(0, 10000 + 1, 2))

print(sys.getsizeof(gen))
print(sys.getsizeof(lst))   
```

### Завдання 10 — Генератор Фібоначчі

Напишіть генератор чисел Фібоначчі й візьміть із нього перші 10 через `itertools.islice`. _(Розділи 6.2–6.3.)_

```python
import itertools


def fibonacci():
    a, b = 0, 1
    while True:
        yield a
        a, b = b, a + b


result = list(itertools.islice(fibonacci(), 10))
print(result)
```

### Завдання 13 — Обробка помилок (try/except/finally)

Напишіть функцію `safe_divide(a, b)`, яка ділить `a` на `b`. Через `try/except` перехопіть `ZeroDivisionError` і поверніть `None` при діленні на нуль. У блоці `finally` виведіть `"обчислення завершено"`. Перевірте на `safe_divide(10, 2)` та `safe_divide(10, 0)`. _(Частина 5.)_

```python
def safe_divide(a, b):
    try:
        return a / b
    except ZeroDivisionError:
        return None
    finally:
        print("обчислення завершено")


print(safe_divide(10, 2))
print(safe_divide(10, 0))
```

### Завдання 14 (⭐) — Декоратор з параметрами

Напишіть декоратор `@repeat(n)`, який виконує обгорнуту функцію `n` разів. Застосуйте його до функції, що друкує повідомлення, і перевірте (не забудьте `functools.wraps`). _(Частина 6.)_

```python
from functools import wraps


def repeat(n):
    def decorator(func):
        @wraps(func)
        def wrapper(*args, **kwargs):
            result = None
            for _ in range(n):
                result = func(*args, **kwargs)
            return result

        return wrapper

    return decorator


@repeat(3)
def greet():
    print("Привіт!")


greet()
```

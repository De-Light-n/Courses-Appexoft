### Завдання 4 — *args

Напишіть функцію `maximum(*args)`, яка повертає найбільший з переданих аргументів (без вбудованого `max`). _(Розділ 1.5.)_

```python
#def maximum(*args):
#    return max(*args)

#print(maximum(30, 56, 78, 34, 809)) 

def maximum(*args):
    largest = 0
    for number in args:
        if number > largest:
            largest = number
    return largest

print(maximum(30, 56, 78, 34, 809)) 
```

```python
def collect(x, into=None):
    if into is None: #конкретно у цьому завданні не дуже зрозуміло, чому не може бути == 
        into = []
    into.append(x)

    return into

collect(1) #появляється список [1]
collect(2) #появляться новий список [2]
new_list = [10]
collect(5, new_list) #into тут не є None, отже створюється не новий список, а використовується заданий та по функції додається значення 5
```

### Завдання 7 — Замикання

Напишіть `make_accumulator()`, що повертає функцію: вона додає передане число до внутрішньої суми й повертає поточну суму. _(Частина 4.)_

```python
def make_accumulator():
    total = 0
    def accum(num):
        nonlocal total
        total += num
        return total
    
    return accum

a = make_accumulator()
print(a(5))
print(a(10))
print(a(1000))
```

### Завдання 8 — Декоратор @timeit

Напишіть декоратор `@timeit`, який друкує час виконання обгорнутої функції (використайте `time.perf_counter`). Не забудьте `functools.wraps`. _(Частина 5.)_

```python
import functools
import time

def timeit(func):
    @functools.wraps(func) 
    def wrapper(*args, **kwargs):
        start = time.perf_counter()

        result = func(*args, **kwargs)

        end = time.perf_counter()

        print(f"час виконання {end - start} секунд")
        return result

    return wrapper

@timeit
def add(a, b):
    return a + b

add(100, 200)
```

### Завдання 9 — Генератор

Напишіть генератор `even_numbers(limit)`, який ліниво видає парні числа до `limit`. Доведіть через `sys.getsizeof`, що він компактніший за відповідний список. _(Частина 6.)_

```python
import sys

def even_numbers(limit):
    for i in range(0, limit + 1):
        if i % 2 == 0:
            yield i

for number in even_numbers(100):
    print(number)

my_list = [i for i in range(100000) if i % 2 == 0]
my_generator = even_numbers(100000)

print(sys.getsizeof(my_list)) #444376 байт
print(sys.getsizeof(my_generator)) #216 байт - займає набагато менше місця

```

### Завдання 10 — Генератор Фібоначчі

Напишіть генератор чисел Фібоначчі й візьміть із нього перші 10 через `itertools.islice`. _(Розділи 6.2–6.3.)_

```python
from itertools import islice

def fibonacci():
    a = 0
    b = 1
    while True:
        yield a
        a, b = b, a + b

result = islice(fibonacci(), 10)
print(list(result))
```

### Завдання 11 — Менеджер контексту

Напишіть менеджер контексту через `@contextmanager`, який друкує `"старт"` на вході та `"кінець"` на виході з блоку `with`. _(Частина 7.)_

```python
from contextlib import contextmanager

@contextmanager
def context():
    print("start")

    yield

    print("end")

with context():
    print('inside')
```

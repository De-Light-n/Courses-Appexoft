### Завдання 1 — Проста функція

### Завдання 3 — **kwargs

Напишіть функцію `describe(**kwargs)`, яка друкує всі передані іменовані параметри у форматі `ключ = значення`. _(Розділ 1.5.)_

```python
def describe(**kwargs):
    for key, value in kwargs.items():
        print(f"{key} = {value}")
    
describe(name = "Ярослав", age = 26, city = "Львів")
```

### Завдання 4 — *args

Напишіть функцію `maximum(*args)`, яка повертає найбільший з переданих аргументів (без вбудованого `max`). _(Розділ 1.5.)_

```python
def maximum(*args):
    if  len(args) == 0:
        return(None)
    max_arg = args[0]
    for arg in args:
        if arg > max_arg:
            max_arg = arg
    return(max_arg)

print(maximum(1, 3, 5, 6, 8, 10, 15, 7))
```

### Завдання 5 — Пастка default

Виправте функцію `def collect(x, into=[]): ...` так, щоб кожен виклик починався з порожнього списку. Доведіть двома викликами. _(Розділ 1.6.)_

```python
def collect(x, into=None):
    if into is None:
        into = []
    into.append(x)
    return into

print(collect("1"))
print(collect("2"))
```

### Завдання 8 — Декоратор @timeit

Напишіть декоратор `@timeit`, який друкує час виконання обгорнутої функції (використайте `time.perf_counter`). Не забудьте `functools.wraps`. _(Частина 5.)_

```python
import time
import functools

def timeit(func):
    @functools.wraps(func)
    def wrapper(*args, **kwargs):
     start_time = time.perf_counter()
     result = func(*args, **kwargs)
     end_time = time.perf_counter()
     exec_time = end_time - start_time 
     print(f"Час виконання:{exec_time} сек.")
     return result
    return wrapper 
    
@timeit
def area(width, height):
    return width * height

a = [1, 2, 4, 6, 8, 14]
b = [2, 4, 1, 7, 3, 43]

for index, (w, h) in enumerate(zip(a, b), start=1):
    result = area(w, h)
    print((f"Площа {index}: {result}"))
```

### Завдання 9 — Генератор

Напишіть генератор `even_numbers(limit)`, який ліниво видає парні числа до `limit`. Доведіть через `sys.getsizeof`, що він компактніший за відповідний список. _(Частина 6.)_

```python
import sys

def even_numbers(limit):
    for number in range(limit + 1):
        if number % 2 == 0:
            yield number 
            
limit = 100000

list_ver = [x for x in range(limit+1) if x % 2 == 0]
gen_ver = even_numbers(limit)

print(f"list_ver size = {sys.getsizeof(list_ver)} bytes")
print(f"gen_ver size = {sys.getsizeof(gen_ver)} bytes")
```

### Завдання 10 — Генератор Фібоначчі

Напишіть генератор чисел Фібоначчі й візьміть із нього перші 10 через `itertools.islice`. _(Розділи 6.2–6.3.)_

```python
import itertools

def fibonachi_gen():
    a = 0
    b = 1
    while True:
        yield a
        c = a
        a = b
        b = c + b

print(list(itertools.islice(fibonachi_gen(),10)))
    
```

### Завдання 11 — Менеджер контексту

Напишіть менеджер контексту через `@contextmanager`, який друкує `"старт"` на вході та `"кінець"` на виході з блоку `with`. _(Частина 7.)_

```python
from contextlib import contextmanager

@contextmanager
def tag():
    print("Старт")
    yield
    print("Фініш")
    
def area(width, height):
    return width * height

a = [1, 2, 4, 6, 8, 14]
b = [2, 4, 1, 7, 3, 43]

with tag():
    for index, (w, h) in enumerate(zip(a, b), start=1):
     result = area(w, h)
     print((f"Площа {index}: {result}"))
```

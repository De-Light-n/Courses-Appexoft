## 1. Що таке Python і як він виконується

Python — інтерпретована мова з **динамічною** та **сильною** типізацією.

- **Інтерпретована**: код не компілюється одразу в машинний код, а виконується інтерпретатором. Насправді CPython (стандартна реалізація) спершу компілює ваш `.py` у **байткод** (`.pyc`), а потім виконує його у віртуальній машині.
- **Динамічна типізація**: тип змінної визначається під час виконання, його не треба оголошувати.
- **Сильна типізація**: Python не робить «мовчазних» дивних приведень типів (наприклад, `1 + "2"` — це помилка, а не `"12"` чи `3`).

Терміни, які варто знати:
- **CPython** — стандартна реалізація мови на C (те, що ви завантажуєте з python.org). Є ще PyPy (з JIT), Jython тощо.
- **GIL (Global Interpreter Lock)** — глобальне блокування, через яке в один момент байткод виконує лише один потік (важливо для теми багатопоточності — повернемось у Модулі 2).

## 2. Змінні: імена й об'єкти (а не «коробки»)

Це фундамент, який відрізняє новачка від того, хто розуміє Python.

У багатьох мовах змінна — це «коробка», в якій лежить значення. **У Python змінна — це ім'я (ярлик), прив'язане до об'єкта в пам'яті.** Оператор `=` не «кладе значення в коробку», а **прив'язує ім'я до об'єкта**.

Кожен об'єкт має:
- **тип** (`type()`),
- **значення**,
- **ідентичність** — унікальний «адрес» у пам'яті (`id()`).

Два імені можуть вказувати на **той самий** об'єкт — це називається *aliasing* (псевдонім).

```python
a = [1, 2, 3]
b = a            # b і a — це ДВА ІМЕНІ одного й того ж об'єкта-списку

print("id(a):", id(a))
print("id(b):", id(b))
print("a is b:", a is b)   # той самий об'єкт?

b.append(4)      # змінюємо об'єкт через ім'я b...
print("a:", a)   # ...і бачимо зміну через ім'я a — бо це ОДИН об'єкт
```

### Наслідок: як передаються аргументи у функції

Python передає аргументи за принципом **«pass by object reference»** (передача посилання на об'єкт). Що станеться всередині функції — залежить від того, **мутабельний** об'єкт чи ні.

```python
def try_to_change(number, my_list):
    number += 10        # int незмінний -> ім'я всередині функції прив'язується до НОВОГО об'єкта
    my_list.append(99)  # list змінний -> МУТУЄМО той самий об'єкт

x = 5
data = [1, 2]
try_to_change(x, data)

print("x залишився:", x)       # 5 — зовнішнє ім'я не змінилось
print("data змінився:", data)  # [1, 2, 99] — об'єкт мутували
```

## 3. Типи даних

### 3.1. Числові типи
- `int` — ціле число **довільної точності** (не обмежене 32/64 бітами!).
- `float` — число з рухомою комою, стандарт IEEE 754 (подвійна точність). Звідси відомі «неточності».
- `bool` — логічний тип. **Важливо: `bool` є підкласом `int`**, тому `True == 1`, `False == 0`.
- `complex` — комплексні числа (`3 + 4j`).

```python
# int довільної точності — жодного переповнення
big = 2 ** 200
print(big)
print("Кількість цифр:", len(str(big)))
```

```python
# float та IEEE 754: класична "неточність"
print(0.1 + 0.2)              # 0.30000000000000004
print(0.1 + 0.2 == 0.3)       # False (!)

# Правильний спосіб порівнювати float:
import math
print(math.isclose(0.1 + 0.2, 0.3))  # True
```

```python
# bool — це підклас int.
print(isinstance(True, int))   # True
print(True + True)             # 2
print(True == 1, False == 0)   # True True
print(sum([True, True, False, True]))  # 3 — інколи так рахують кількість True
```

### 3.2. Рядки, `None` та контейнери (оглядово)
- `str` — незмінний рядок Unicode (детально — наступна лекція).
- `NoneType` — єдиний об'єкт `None`, «відсутність значення».
- Контейнери: `list`, `tuple`, `dict`, `set` — детально наступної лекції.

### 3.3. Mutable vs Immutable — критично важливо

| Незмінні (immutable) | Змінні (mutable) |
|---|---|
| `int`, `float`, `bool`, `str`, `tuple`, `frozenset`, `bytes` | `list`, `dict`, `set`, `bytearray` |

Незмінний об'єкт **не можна змінити після створення** — будь-яка «зміна» насправді створює новий об'єкт.

```python
# Доказ незмінності рядка: id змінюється при "зміні"
s = "abc"
print("до:", id(s))
s += "d"                 # це НЕ зміна рядка, а створення нового об'єкта
print("після:", id(s))   # інший id!

# А список мутується "на місці" — id той самий
lst = [1, 2]
print("до:", id(lst))
lst.append(3)
print("після:", id(lst)) # той самий id
```

## 4. `==` проти `is`

- **`==`** порівнює **значення** (викликає метод `__eq__`). «Чи однакові вони на вигляд?»
- **`is`** порівнює **ідентичність** — чи це **той самий об'єкт** у пам'яті (порівнює `id()`). «Це буквально одне й те саме?»

Правило: **`is` — тільки для `None`, `True`, `False`** (синглтони). Для порівняння значень завжди `==`.

```python
a = [1, 2, 3]
b = [1, 2, 3]   # ОКРЕМИЙ об'єкт з таким самим вмістом

print("a == b:", a == b)   # True  -> вміст однаковий
print("a is b:", a is b)   # False -> це різні об'єкти в пам'яті
print("id(a) != id(b):", id(a) != id(b))
```

### Чому `256 is 256`, а `257 is 257` буває `False`?

CPython **кешує малі цілі числа** від **−5 до 256** (вони створюються один раз при старті інтерпретатора). Тому два імені зі значенням `256` вказують на **той самий** закешований об'єкт. А `257` вже поза кешем — створюються два різні об'єкти.

⚠️ Це **деталь реалізації CPython**, а не гарантія мови. Ба більше: якщо писати числа-літерали в **одному рядку/комірці**, компілятор може «згорнути» їх в один об'єкт навіть для великих чисел — тому нижче ми навмисно створюємо числа через `int("...")`, щоб побачити саме кеш. Висновок один: на `is` для чисел і рядків покладатись **не можна**, тільки `==`.

```python
# Створюємо числа під час виконання (int("...")), щоб побачити саме КЕШ,
# а не оптимізацію компілятора над літералами.
a = int("256")
b = int("256")
print("256:", a is b)   # True  -> закешоване мале число (-5..256)

a = int("257")
b = int("257")
print("257:", a is b)   # False -> поза діапазоном кешу

# Те саме зі стрічками — "string interning".
x = "hello"                       # літерал, схожий на ідентифікатор -> інтернується
y = "hello"
print("hello (літерали):", x is y)          # True

z = "".join(["h", "e", "l", "l", "o"])       # той самий текст, але зібраний під час виконання
print("hello (runtime):  ", x is z)          # False -> це вже інший обʼєкт
print("а == усе одно:    ", x == z)          # True  -> значення однакове
```

```python
# ПРАВИЛЬНЕ використання is — тільки для синглтонів None/True/False
value = None
print(value is None)        # True — так треба перевіряти на None
print(value is not None)    # зручний зворотний варіант
```

## 5. Оператори

### 5.1. Групи операторів
- **Арифметичні:** `+ - * / // % **`
- **Порівняння:** `== != > < >= <=` (можна ланцюжком: `1 < x < 10`)
- **Логічні:** `and`, `or`, `not` — з коротким обчисленням (short-circuit)
- **Побітові:** `& | ^ ~ << >>`
- **Ідентичності:** `is`, `is not`
- **Належності:** `in`, `not in`

```python
# --- Арифметичні оператори (докладно) ---
print(7 + 3)    # 10  додавання
print(7 - 3)    # 4   віднімання
print(7 * 3)    # 21  множення
print(7 / 2)    # 3.5   -> звичайне ділення ЗАВЖДИ дає float (навіть 4 / 2 == 2.0)
print(7 // 2)   # 3     -> цілочисельне ділення (відкидає дробову частину)
print(7 % 2)    # 1     -> остача від ділення (класика для перевірки на парність)
print(2 ** 10)  # 1024  -> піднесення до степеня

# // і % з відʼємними числами часто дивують:
print(-7 // 2)  # -4  округлення ВНИЗ (до меншого), а не до нуля
print(-7 % 2)   # 1   знак результату збігається зі знаком дільника

# divmod дає одразу частку і остачу
print(divmod(17, 5))   # (3, 2)
```

```python
# --- Оператори порівняння: завжди повертають bool ---
print(5 > 3)    # True
print(5 == 5)   # True   рівність значень
print(5 != 4)   # True   нерівність
print(3 <= 3, 3 >= 4)   # True False

# --- Логічні оператори ---
print(True and False)   # False  -> and: True лише якщо ОБИДВА True
print(True or False)    # True   -> or:  True якщо хоча б один True
print(not True)         # False  -> not: інвертує

# --- Оператори присвоєння (augmented) — скорочення для "змінити й записати назад" ---
x = 10
x += 5    # те саме, що x = x + 5
x *= 2    # x = x * 2
x //= 3   # x = x // 3
print(x)  # 10
```

```python
# Ланцюгове порівняння — так пишуть тільки в Python
x = 7
print(1 < x < 10)          # еквівалент (1 < x) and (x < 10)

# Побітові оператори
print(5 & 3)   # 1  (101 & 011 = 001)
print(5 | 2)   # 7
print(1 << 4)  # 16 (зсув вліво = множення на 2**4)

# Належність
print("а" in "Львів")      # True
print(3 in [1, 2, 3])       # True
```

### 5.2. Short-circuit (коротке обчислення) — часте питання

- `A and B`: якщо `A` хибне — `B` **навіть не обчислюється**, результат = `A`.
- `A or B`: якщо `A` істинне — `B` не обчислюється, результат = `A`.

Важливо: `and`/`or` повертають **не `True`/`False`, а один з операндів**. Це використовують для значень за замовчуванням.

```python
def loud():
    print("  loud() викликана!")
    return True

print("--- False and loud() ---")
print(False and loud())   # loud() НЕ викликається -> надрукує False

print("--- True or loud() ---")
print(True or loud())     # loud() НЕ викликається -> надрукує True

# and/or повертають операнд, а не bool:
name = ""
print(name or "Гість")    # "Гість" — типовий трюк для значень за замовчуванням
print(0 or 42)            # 42
```

### 5.3. Truthiness — «правдивість» об'єктів

У логічному контексті (`if`, `while`, `and`, `or`) будь-який об'єкт зводиться до `True`/`False`.

**Хибні (falsy):** `False`, `None`, `0`, `0.0`, `""`, `[]`, `{}`, `()`, `set()`.
**Правдиві (truthy):** майже все інше (непорожні контейнери, будь-які ненульові числа).

```python
for obj in [0, 1, "", "текст", [], [1], None, {}]:
    print(f"{obj!r:>8}  ->  {bool(obj)}")

# Тому перевіряють так, а не len(x) == 0:
items = []
if not items:
    print("Список порожній")
```

## ⚡ Каверзні питання (з відповідями)

Спробуйте спершу відповісти самі, потім запустіть код.

**Q1. Що поверне `0.1 + 0.2 == 0.3` і чому?**
`False`. `float` зберігається у двійковому IEEE 754, і `0.1`, `0.2`, `0.3` не мають точного двійкового представлення. Порівнювати треба через `math.isclose()`.

**Q2. `a is b` для двох списків `[1,2,3]` — True чи False?**
`False`. `is` порівнює ідентичність (той самий об'єкт), а це два різні об'єкти. `==` дасть `True`.

**Q3. Чому `a = 256; b = 256; a is b` → True, а для `257` → False?**
CPython кешує малі цілі `-5..256`. Для `257` створюються два окремих об'єкти. Це деталь реалізації — на `is` для чисел покладатись не можна.

**Q4. Чому `True == 1` і `isinstance(True, int)` дають True?**
`bool` — підклас `int`. `True` дорівнює `1`, `False` — `0`. Тому `True + True == 2`.

```python
# Q5. Що надрукується? (мутабельність + aliasing)
a = [1, 2, 3]
b = a
b = b + [4]        # УВАГА: b + [4] створює НОВИЙ список, b перепривʼязується
print("a:", a)     # [1, 2, 3] — не змінився!
print("b:", b)     # [1, 2, 3, 4]

# А тут — інакше:
a = [1, 2, 3]
b = a
b += [4]           # += для списку = list.extend -> МУТАЦІЯ того самого об'єкта
print("a:", a)     # [1, 2, 3, 4] — змінився! (a і b — один обʼєкт)
print("b:", b)
```

**Q5 (пояснення).** `b = b + [4]` створює **новий** об'єкт і перепривʼязує `b`, тому `a` не змінюється. А `b += [4]` для списку — це `list.__iadd__` (по суті `extend`), який **мутує той самий** об'єкт, тому `a` теж «бачить» зміну. Різниця `+` vs `+=` для mutable — класична пастка.

```python
# Q6. Порожній рядок, 0 і None — чи вони "рівні" в if?
print(bool(""), bool(0), bool(None))   # усі False у логічному контексті
print("" == 0)     # False! (falsy != рівність)
print(None == False)  # False
```

**Q6 (пояснення).** «Falsy» (зводиться до `False` в `if`) — це **не те саме**, що «дорівнює `False`/`0`». `""`, `0`, `None` усі falsy, але `"" == 0` → `False`, `None == False` → `False`.

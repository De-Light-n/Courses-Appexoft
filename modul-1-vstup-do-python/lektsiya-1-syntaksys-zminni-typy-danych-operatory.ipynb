## 1. Що таке Python

**Python** — одна з найпопулярніших мов програмування у світі. Її люблять за те, що код
читається майже як звичайна англійська, а писати можна швидко.

Писати код можна вже з цим розумінням. Але **як саме Python виконує ваш код** — окрема важлива

## 1.1. 🔎 Важливо знати: як Python виконує ваш код

На старті це не потрібно, щоб писати програми. Але це **класичне питання співбесіди**, тож
розберемо простими словами — і одразу побачимо на прикладі.

**Інтерпретована чи компільована мова?** Насправді потроху обидва. Коли ви запускаєте `.py`,
стандартна реалізація — **CPython** — спершу **компілює** код у проміжний **байткод**
(ті самі файли `.pyc` у теці `__pycache__`), а потім виконує його у **віртуальній машині Python
(PVM)**. Тобто справжнього машинного коду, як у C, не створюється.

**Динамічна типізація** — тип змінної визначається під час виконання; його не оголошують
заздалегідь. Можна навіть `x = 5`, а потім `x = "текст"`.

**Сильна типізація** — Python не робить «мовчазних» дивних перетворень. `1 + "2"` — це **помилка**,
а не `"12"` чи `3` (як буває в деяких інших мовах).

**Реалізації мови:** **CPython** (стандартна, написана на C — та, що з python.org), **PyPy**
(швидша, з JIT-компіляцією), **Jython**, **IronPython**.

**GIL (Global Interpreter Lock)** — глобальне блокування в CPython, через яке в один момент часу
**байткод виконує лише один потік**. Наслідок: багатопотоковість **не пришвидшує** обчислювальні
(CPU-bound) задачі. Це часте питання співбесіди; детальніше — у лекції про функції та паралелізм.

````python
# Побачимо все на власні очі. import підключає готові модулі.
import sys, dis

print("Реалізація Python:", sys.implementation.name)   # напр. 'cpython'
print("Версія:", sys.version.split()[0])
````

А тепер найцікавіше — заглянемо у **байткод**. Функція `dis.dis()` показує, у які інструкції
CPython перетворив наш код. Саме ці інструкції й виконує віртуальна машина:

````python
import dis

def add(a, b):
    return a + b

# Дизасемблюємо функцію -> побачимо інструкції байткоду (LOAD_FAST, BINARY_ADD, RETURN_VALUE...)
dis.dis(add)
````

> 💡 Не треба вчити ці інструкції напам'ять. Головне — **розуміти ідею**: ваш зрозумілий код
> Python спершу перетворює на такий низькорівневий байткод, а вже його виконує PVM.

## 4. Змінні

**Змінна** — це **ім'я**, яке ми даємо якомусь значенню, щоб потім до нього звертатися.
Уявіть підписаний контейнер: ми кладемо в нього значення й пишемо на ньому назву.

Створюють змінну через знак `=` (він читається як «присвоїти», а **не** «дорівнює»):

```
ім'я = значення
```

````python
# Створюємо змінні
name = "Олена"      # у змінну name поклали текст
age = 25            # у змінну age поклали число

# Тепер можемо звертатися до значень за іменами
print(name)
print(age)
print("Ім'я:", name, "| Вік:", age)
````

Значення змінної можна **міняти** — тому вона й «змінна». Нове значення просто замінює старе.

````python
balance = 100
print("Було:", balance)

balance = 150          # поклали нове значення — старе (100) зникло
print("Стало:", balance)

# Змінну можна використати, щоб порахувати нове її ж значення
balance = balance + 50 # взяли поточне (150), додали 50, записали назад
print("Після поповнення:", balance)   # 200
````

### Правила іменування змінних
- лише **літери, цифри та `_`** (нижнє підкреслення);
- **не можна починати з цифри** (`2name` — помилка, `name2` — можна);
- **регістр важливий**: `age`, `Age` та `AGE` — це три **різні** змінні;
- ім'я має бути **зрозумілим**: `price`, `user_name` — добре; `x`, `a1` — погано (крім зовсім простих випадків).

За домовленістю в Python імена змінних пишуть **малими літерами**, а слова розділяють `_`
(це називається `snake_case`): `user_name`, `total_price`, `is_active`.

````python
# Приклади коректних імен
user_name = "admin"
total_price = 499.99
is_active = True
print(user_name, total_price, is_active)

# Регістр важливий — це різні змінні:
color = "червоний"
Color = "синій"
coLor = "Жовтий"
print(color)   # червоний
print(Color)   # синій
print(coLor)   # Жовтий
````

## 5. Типи даних

Кожне значення в Python має свій **тип** — тобто «сорт» даних. Від типу залежить, що з даними
можна робити. Розгляньмо базові типи по черзі — **кожен окремо**.

### 5.1. `int` — цілі числа

`int` (від *integer*) — це **ціле число**: `0`, `7`, `-42`, `1000000`. Без дробової частини.
З цілими числами можна робити всю звичну арифметику.

````python
apples = 10
oranges = 5

print("Всього фруктів:", apples + oranges)   # 15
print("Різниця:", apples - oranges)          # 5
print("Добуток:", apples * oranges)          # 50

# Цікава особливість Python: цілі числа можуть бути ЯК ЗАВГОДНО великими
huge = 2 ** 100      # 2 у сотому степені
print("Дуже велике число:", huge)
````

### 5.2. `float` — дробові числа

`float` — це число **з крапкою** (дробове): `3.14`, `0.5`, `-2.0`.
Зверніть увагу: розділювач — **крапка**, а не кома.

````python
price = 19.99
discount = 0.1        # 10%

print("Ціна:", price)
print("Знижка у грошах:", price * discount)   # 1.999
print("Ціна зі знижкою:", price - price * discount)

# Якщо поділити одне число на інше, результат ЗАВЖДИ буде float (з крапкою):
print(10 / 2)         # 5.0, а не просто 5
````

> ⚠️ **Маленька дивина, яку варто знати заздалегідь.** Комп'ютер зберігає дробові числа
> наближено, тому інколи бувають «хвости»:

````python
print(0.1 + 0.2)      # 0.30000000000000004, а не рівно 0.3

# Це нормально для всіх мов програмування, не тільки Python.
# Коли треба порівнювати дробові — користуються спеціальним інструментом.
# import підключає готовий МОДУЛЬ (набір функцій). Тут це модуль math.
import math
print(math.isclose(0.1 + 0.2, 0.3))   # True — "практично рівні"
````

### 5.3. `str` — рядки (текст)

`str` (від *string*) — це **текст**: будь-які символи в лапках. Це можуть бути імена, повідомлення,
адреси тощо. Рядкам буде присвячена **вся наступна лекція**, тут — лише основи.

````python
greeting = "Привіт"
name = "Олено"

# Рядки можна "додавати" (склеювати) знаком +
message = greeting + ", " + name + "!"
print(message)        # Привіт, Олено!

# Рядок можна повторити знаком *
print("ха" * 3)       # хахаха

# Довжину рядка (кількість символів) дає функція len()
print(len("Python"))  # 6
````

### 5.4. `bool` — логічний тип (правда/неправда)

`bool` (від *boolean*) має лише **два значення**: `True` (істина) і `False` (хиба).
Пишуться саме так — з великої літери. Цей тип — основа для перевірок і умов.

````python
is_raining = True
is_sunny = False

print(is_raining)     # True
print(is_sunny)       # False

# Найчастіше bool з'являється як результат ПОРІВНЯННЯ:
print(10 > 5)         # True  — правда, що 10 більше за 5
print(3 == 7)         # False — неправда, що 3 дорівнює 7
````

### 5.5. `None` — «нічого / значення відсутнє»

`None` — це особливе значення, яке означає **«нічого немає»** або «значення поки не задано».
Воно одне на весь Python. Не плутайте його з `0` чи порожнім рядком `""` — це саме «порожнеча».

````python
winner = None         # переможець ще не відомий
print(winner)         # None

# Пізніше можемо задати справжнє значення:
winner = "Команда А"
print(winner)         # Команда А
````

## 6. Як дізнатися тип: `type()`

Функція `type()` показує, якого типу значення чи змінна. Дуже корисно, коли щось не працює
й треба зрозуміти, з чим саме ви маєте справу.

````python
print(type(42))          # <class 'int'>   — ціле число
print(type(3.14))        # <class 'float'> — дробове
print(type("текст"))     # <class 'str'>   — рядок
print(type(True))        # <class 'bool'>  — логічне
print(type(None))        # <class 'NoneType'>

# type() працює і зі змінними:
age = 25
print(type(age))         # <class 'int'>
````

## 7. Перетворення типів

Іноді потрібно **перетворити** значення з одного типу на інший. Для цього є функції з назвами типів:
`int()`, `float()`, `str()`, `bool()`.

Навіщо це треба? Наприклад, коли користувач вводить число — Python отримує його **як текст**,
і щоб порахувати, текст спершу треба перетворити на число.

````python
# Текст -> число
text_number = "100"          # це РЯДОК, не число (у лапках)
real_number = int(text_number)
print(real_number + 50)      # 150 — тепер можна рахувати

# Число -> текст (щоб склеїти з іншим рядком)
age = 25
print("Мені " + str(age) + " років")   # str() перетворює 25 на "25"

# Ціле -> дробове й навпаки
print(float(7))     # 7.0
print(int(3.9))     # 3 — увага: НЕ округлює, а ВІДКИДАЄ дробову частину!
````

> ⚠️ Перетворити можна не завжди. `int("привіт")` дасть помилку, бо «привіт» — це не число.
> Це нормально: Python попереджає, що так робити не можна.
>
> ℹ️ У прикладі нижче ми вперше бачимо `try`/`except` — це спосіб **«перехопити» помилку**,
> щоб програма не зупинилась. Повноцінно розберемо його в наступному модулі; поки достатньо
> знати, що так помилку можна впіймати й обробити.

````python
# Приклад помилки при перетворенні (обгорнули в try, щоб зошит не зупинився)
try:
    int("привіт")
except ValueError as e:
    print("Помилка:", e)
````

## 9. Оператори

**Оператор** — це символ, який виконує дію над значеннями (наприклад, `+` додає).
Розгляньмо основні групи операторів окремо.

### 9.1. Арифметичні оператори

````python
print(7 + 3)     # 10   додавання
print(7 - 3)     # 4    віднімання
print(7 * 3)     # 21   множення
print(7 / 2)     # 3.5  ділення (результат завжди дробовий, з крапкою)
print(7 // 2)    # 3    цілочисельне ділення (скільки РАЗІВ вміщається, без остачі)
print(7 % 2)     # 1    остача від ділення (що ЛИШИЛОСЬ)
print(2 ** 10)   # 1024 піднесення до степеня (2 у 10-му степені)
````

Два оператори часто плутають новачки, тож розберемо детальніше:

- **`//` (цілочисельне ділення)** — ділить і **відкидає** дробову частину. `17 // 5` = `3`
  (пʼятірка вміщається в 17 рівно 3 рази).
- **`%` (остача, «модуло»)** — показує, що **лишилось** після цілого ділення. `17 % 5` = `2`.

Остача `%` — дуже корисна. Наприклад, число **парне**, якщо остача від ділення на 2 дорівнює 0:

````python
number = 17
print("17 // 5 =", 17 // 5)   # 3  — цілих трійок
print("17 % 5  =", 17 % 5)    # 2  — лишок

# Перевірка на парність через остачу від ділення на 2:
print("10 парне? ", 10 % 2 == 0)   # True
print("7 парне?  ", 7 % 2 == 0)    # False
````

### 9.2. Оператори порівняння

Порівнюють два значення й **завжди повертають `True` або `False`**.
Зверніть увагу: рівність — це **подвійний** знак `==` (одинарний `=` — це присвоєння!).

````python
print(5 > 3)     # True   більше
print(5 < 3)     # False  менше
print(5 >= 5)    # True   більше або дорівнює
print(5 <= 4)    # False  менше або дорівнює
print(5 == 5)    # True   ДОРІВНЮЄ (два знаки =)
print(5 != 3)    # True   НЕ дорівнює

# Порівнювати можна не тільки числа, а й рядки:
print("яблуко" == "яблуко")   # True
print("яблуко" == "груша")    # False
````

> 🧠 Найчастіша помилка новачка — переплутати `=` та `==`.
> `=` — «поклади значення у змінну». `==` — «чи рівні ці два значення?».

### 9.3. Логічні оператори: `and`, `or`, `not`

Дозволяють **поєднувати** кілька умов:
- **`and`** (і) — `True`, лише якщо **обидві** умови істинні;
- **`or`** (або) — `True`, якщо **хоча б одна** умова істинна;
- **`not`** (не) — **перевертає** значення: `not True` = `False`.

````python
age = 20
has_ticket = True

# and: обидві умови мають бути True
print(age >= 18 and has_ticket)   # True — і повнолітній, і має квиток

# or: достатньо однієї
print(age < 12 or age >= 65)      # False — ні дитина, ні пенсіонер

# not: перевертає
print(not has_ticket)             # False
````

Таблиця для `and` та `or` (щоб було наочно):

````python
print("--- and (і) ---")
print(True  and True)    # True
print(True  and False)   # False
print(False and False)   # False

print("--- or (або) ---")
print(True  or False)    # True
print(False or True)     # True
print(False or False)    # False
````

## 10. «Правдивість» значень (truthiness)

У Python **будь-яке** значення можна перевірити на «правдивість» — тобто чи вважається воно
`True` чи `False`, коли потрапляє в умову. Це знадобиться далі, коли вивчатимемо `if`.

Правило просте — **«порожнє / нульове» вважається `False`**, а решта — `True`:
- **Falsy (хибні):** `False`, `None`, число `0`, порожній рядок `""`, порожній список `[]`.
- **Truthy (правдиві):** будь-яке інше число, непорожній текст, непорожній список.

Перевірити, як значення поводиться в умові, можна функцією `bool()`:

````python
# Хибні (falsy) значення:
print(bool(0))       # False
print(bool(""))      # False — порожній рядок
print(bool(None))    # False

# Правдиві (truthy) значення:
print(bool(42))      # True
print(bool("текст")) # True — будь-який непорожній текст
print(bool(-1))      # True — будь-яке ненульове число
````

## 11. `==` проти `is` — порівняння значень і об'єктів

Це одне з **найчастіших питань на співбесідах** — розберемо уважно, бо плутанина тут
породжує підступні баги.

Є два схожі, але **різні** способи порівняння:
- **`==`** запитує: «**чи однакові значення?**» (те, що бачить людина). Використовуйте **майже завжди**.
- **`is`** запитує: «**чи це буквально той самий об'єкт** у пам'яті?». Потрібне **лише** для перевірки на `None`.

````python
a = [1, 2, 3]
b = [1, 2, 3]     # окремий список з таким самим вмістом

print(a == b)     # True  — значення однакові
print(a is b)     # False — це РІЗНІ об'єкти (просто схожі)

# Єдине правильне застосування is — перевірка на None:
value = None
print(value is None)      # True — саме так перевіряють на "порожнечу"
````

**Запам'ятайте просте правило:** для звичайних порівнянь завжди беріть `==`,
а `is` використовуйте **тільки** з `None` (`if x is None:`).

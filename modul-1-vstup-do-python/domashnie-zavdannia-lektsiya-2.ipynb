### Завдання 1 — Оцінка за балом

Задайте змінну `score` (0–100). За допомогою `if/elif/else` виведіть оцінку: `A` (90+), `B` (75+), `C` (60+), інакше `F`. _(Розділ 1.3.)_

```python
score = 69

if score >= 90:
    print('A')
elif score >= 75:
    print('B')
elif score >= 60:
    print('C')
else:
    print('F')
```

### Завдання 2 — Парне чи непарне

Запитайте число через `input()`, перетворіть на `int` і виведіть, парне воно чи непарне (підказка: `% 2`). _(Розділи 1.1–1.2.)_

```python
a = int(input('Введіть число: '))
if a % 2 == 0:
    print('Парне')
else:
    print("Непарне")
```

### Завдання 3 — Тернарний оператор

Дано `age`. Одним рядком (тернарний оператор) запишіть у змінну `status` значення `"дорослий"` або `"дитина"` і виведіть його. _(Розділ 1.5.)_

```python
age = 67

status = 'Дорослий' if age >= 18 else 'дитина'
print(status)
```

### Завдання 4 — Сума чисел

За допомогою циклу `for` та `range` порахуйте суму всіх чисел від 1 до 100. Виведіть результат. _(Розділ 2.1.)_

```python
total = 0
for i in range(0, 101):
    total += i
    
print(total)
```

### Завдання 5 — Таблиця множення

Виведіть таблицю множення числа 7 (від 7×1 до 7×10) за допомогою циклу `for`. Формат: `7 x 1 = 7`. _(Розділи 2.1.)_

```python
for i in range (1, 10):
    print(f"7 * {i} = {7 * i}")
```

### Завдання 6 — while

За допомогою циклу `while` виведіть числа від 10 до 1 у зворотному порядку. _(Розділ 2.2.)_

```python
n = 10
while n >= 1:
    print(n)
    n -= 1
```

### Завдання 7 — break

Пройдіть циклом по списку `[5, 12, 8, 20, 3]` і виведіть перше число, більше за 10, після чого зупиніть цикл через `break`. _(Розділ 2.3.)_

```python
for n in [5, 12, 8, 20, 3]:
    if n > 10:
        print(f'Число більше за 10: {n}')
        break   
```

### Завдання 8 — continue

Виведіть усі числа від 1 до 20, ПРОПУСКАЮЧИ ті, що діляться на 3 (через `continue`). _(Розділ 2.3.)_

```python
for i in range(1, 21):
    if i % 3 == 0:
        continue
    print(i)
```

### Завдання 9 — enumerate

Дано список `["кава", "чай", "сік"]`. Виведіть його з нумерацією у форматі `1) кава`, `2) чай`, ... через `enumerate`. _(Розділ 2.4.)_

```python
napiy = ['кава', 'чай', "сік"]

for index, item in enumerate(napiy, start=1):
    print(f'{index}) {item}')
```

### Завдання 10 — zip

Дано два списки: `products = ["хліб", "молоко"]` та `prices = [25, 32]`. За допомогою `zip` виведіть `хліб — 25 грн`, `молоко — 32 грн`. _(Розділ 2.5.)_

```python
products = ['хліб', 'молоко']
prices = ['25', '32']

for prod, price in zip(products, prices):
    print(f'{prod} - {price} грн.')
```

### Завдання 11 (⭐) — for/else

Напишіть цикл, що шукає в списку `[4, 8, 15, 16, 23]` число, яке ділиться на 7. Якщо знайшли — виведіть його і `break`; якщо ні — блок `else` повідомляє, що такого немає. _(Розділ 2.6.)_

```python
for number in [4, 8, 15, 16, 23]:
    if number % 7 == 0:
        print(f'Число {number} ділиться націло на 7')
        break
else:
    print('Такого числа немає')
```

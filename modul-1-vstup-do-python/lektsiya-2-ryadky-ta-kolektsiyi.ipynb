## 1. Рядки (`str`)

`str` — **незмінна** послідовність символів Unicode. Кожен «символ» — теж рядок довжини 1 (окремого типа `char` немає).

```python
s = "Львів"
print(len(s))          # довжина
print(s[0], s[-1])     # індексація: перший і останній символ
print(s[1:3])          # зріз [start:stop]
print(s[::-1])         # розворот рядка через зріз з кроком -1

# Незмінність: не можна змінити символ на місці
try:
    s[0] = "л"
except TypeError as e:
    print("TypeError:", e)
```

### 1.1. Зрізи (slicing) — `[start:stop:step]`
Працюють для всіх послідовностей (рядки, списки, кортежі). `stop` **не включається**.

```python
word = "Python"
print(word[0:3])    # 'Pyt'  -> індекси 0,1,2
print(word[2:])     # 'thon' -> від 2 до кінця
print(word[:2])     # 'Py'   -> від початку до 2
print(word[::2])    # 'Pto'  -> кожен другий
print(word[::-1])   # 'nohtyP' -> реверс
```

### 1.2. Корисні методи рядків

```python
text = "  Python — Чудова Мова  "
print(text.strip())                 # прибрати пробіли по краях
print(text.strip().lower())         # у нижній регістр
print(text.strip().upper())         # у верхній регістр
print(text.replace("Чудова", "Потужна"))
print("a,b,c".split(","))           # рядок -> список
print("-".join(["2024", "07", "02"]))  # список -> рядок
print("Python".startswith("Py"), "Python".endswith("on"))
print("Ціна 100 грн".find("грн"))   # індекс підрядка (або -1)
print("hello".center(11, "*"))       # *** hello ***
```

### 1.3. Форматування: f-strings (рекомендовано)

f-рядки (`f"..."`, з Python 3.6) — найзручніший і найшвидший спосіб. У `{}` можна вставляти будь-який вираз і **специфікатори формату**.

```python
name, price = "Ноутбук", 24999.5
print(f"{name}: {price} грн")
print(f"{name}: {price:.2f} грн")     # 2 знаки після коми
print(f"{price:,.2f}")                # роздільник тисяч -> 24,999.50
print(f"{0.1234:.1%}")                # відсотки -> 12.3%
print(f"{42:>6}")                     # вирівнювання праворуч у полі 6
print(f"{name=}")                     # debug-синтаксис (3.8+) -> name='Ноутбук'
```

### 1.4. `str` vs `bytes` та Unicode — часте питання

- `str` — це **текст** (послідовність Unicode-символів).
- `bytes` — це **байти** (сирі дані: файли, мережа).
- Перетворення: `str.encode()` → `bytes`, `bytes.decode()` → `str`. Кодування за замовчуванням — **UTF-8**.

Українська літера в UTF-8 займає **2 байти**, тому «довжина в символах» ≠ «довжина в байтах».

```python
s = "Привіт"
b = s.encode("utf-8")        # str -> bytes
print(b)                     # b'\xd0\x9f...'
print("символів:", len(s))   # 6
print("байтів:", len(b))     # 12 (кирилиця = 2 байти/символ у UTF-8)
print("назад у str:", b.decode("utf-8"))
```

## 2. Колекції: `list`, `tuple`, `set`, `dict`

| Тип | Впорядкований | Змінний | Дублікати | Синтаксис | Коли обирати |
|---|---|---|---|---|---|
| `list` | так | так | так | `[1, 2]` | послідовність, що змінюється |
| `tuple` | так | **ні** | так | `(1, 2)` | незмінний набір, ключ dict |
| `set` | ні | так | **ні** | `{1, 2}` | унікальність, швидка перевірка `in` |
| `dict` | так* | так | ключі унікальні | `{"a": 1}` | пари ключ→значення |

\* `dict` зберігає порядок вставки, **гарантовано з Python 3.7**.

### 2.1. `list` — список

```python
nums = [3, 1, 2]
nums.append(4)         # додати в кінець
nums.insert(0, 0)      # вставити за індексом
nums.remove(1)         # видалити за значенням (перше входження)
popped = nums.pop()    # видалити й повернути останній
print(nums, "видалено:", popped)

nums.sort()            # сортує НА МІСЦІ (мутує список, повертає None!)
print("sort():", nums)
print("sorted():", sorted([5, 3, 9], reverse=True))  # повертає новий список
print("len / max / sum:", len(nums), max(nums), sum(nums))
```

```python
# ⚠️ Класична пастка: sort() повертає None
result = [3, 1, 2].sort()
print("result =", result)   # None! sort() мутує на місці й повертає None
# правильно: використати sorted() або спершу відсортувати, потім використати
```

### 2.2. `tuple` — кортеж (незмінний список)

Незмінний → може бути **ключем словника** та елементом `set`. Часто використовують для «записів» і для повернення кількох значень із функції.

```python
point = (10, 20)
x, y = point                 # розпакування (unpacking)
print("x =", x, "y =", y)

# Кортеж з одного елемента ОБОВʼЯЗКОВО з комою:
not_a_tuple = (5)            # це просто int 5
real_tuple = (5,)            # а це кортеж
print(type(not_a_tuple), type(real_tuple))

# Розширене розпакування зірочкою
first, *middle, last = [1, 2, 3, 4, 5]
print(first, middle, last)   # 1 [2, 3, 4] 5
```

### 2.3. `set` — множина

Невпорядкована колекція **унікальних** елементів. Головні переваги: миттєве прибирання дублікатів і **дуже швидка перевірка належності** `x in set` (у середньому O(1) проти O(n) для списку).

```python
nums = [1, 2, 2, 3, 3, 3]
unique = set(nums)
print("унікальні:", unique)

a = {1, 2, 3}
b = {3, 4, 5}
print("обʼєднання a | b:", a | b)         # {1,2,3,4,5}
print("перетин   a & b:", a & b)          # {3}
print("різниця   a - b:", a - b)          # {1,2}
print("симетрична a ^ b:", a ^ b)         # {1,2,4,5}

# set містить лише хешовані (незмінні) елементи:
try:
    bad = {[1, 2]}          # список не хешований
except TypeError as e:
    print("TypeError:", e)
```

## 3. `dict` — словник, і як він працює всередині ⭐

`dict` зберігає пари **ключ → значення** й дає доступ до значення за ключем у **середньому за O(1)**. Це найважливіша структура даних у Python (навіть простір імен модулів — це dict).

```python
user = {"name": "Ірина", "age": 25}
print(user["name"])              # доступ за ключем
print(user.get("email", "—"))    # безпечний доступ зі значенням за замовчуванням

user["email"] = "iryna@mail.com" # додати/оновити
del user["age"]                  # видалити ключ

for key, value in user.items():  # ітерація по парах
    print(f"  {key}: {value}")

print("keys:", list(user.keys()))
print("values:", list(user.values()))
```

### 3.1. Як `dict` працює всередині (хеш-таблиця)

1. Коли ви кладете `d[key] = value`, Python рахує **`hash(key)`** — ціле число.
2. За хешем визначається «комірка» (bucket) у внутрішньому масиві, куди покласти пару.
3. Пошук `d[key]` рахує той самий хеш → одразу йде в потрібну комірку → **O(1)** у середньому.
4. Якщо два ключі потрапили в одну комірку (**колізія**), Python порівнює їх через `==` і шукає далі за спеціальною схемою (open addressing).

Наслідки, які треба розуміти:
- **Ключем може бути лише хешований (hashable) об'єкт.** Хешовані = незмінні: `str`, `int`, `float`, `bool`, `tuple` (з хешованих елементів), `frozenset`. `list`, `dict`, `set` — **не** можна.
- Правило: якщо `a == b`, то `hash(a) == hash(b)` (інакше словник «загубить» ключ).
- У найгіршому випадку (усі колізії) складність деградує до O(n), але на практиці — O(1).

```python
# hash() працює для незмінних; для змінних — помилка
print("hash('abc'):", hash("abc"))
print("hash((1,2)):", hash((1, 2)))

try:
    hash([1, 2])                 # список не хешований -> не може бути ключем
except TypeError as e:
    print("TypeError:", e)

# Тому tuple можна за ключ, а list — ні:
coords = {(50.45, 30.52): "Київ", (49.84, 24.03): "Львів"}
print(coords[(49.84, 24.03)])
```

```python
# ⭐ Дуже каверзно: True і 1 — це "один ключ", бо hash(True)==hash(1) і True==1
d = {}
d[1] = "ціле"
d[True] = "булеве"
d[1.0] = "дробове"
print(d)               # {1: 'дробове'} — усі три "злилися" в один ключ!
print(hash(1), hash(True), hash(1.0))  # усі однакові -> 1
```

### 3.2. Порядок вставки та зручні методи

```python
from collections import Counter, defaultdict

# Порядок вставки зберігається (гарантія з 3.7)
d = {}
for ch in "bca":
    d[ch] = ord(ch)
print("порядок:", list(d))   # ['b', 'c', 'a'] — як вставляли

# Counter — підрахунок частоти
print(Counter("mississippi"))

# defaultdict — автозначення за замовчуванням, зручно для групування
groups = defaultdict(list)
for word in ["apple", "avocado", "banana", "cherry"]:
    groups[word[0]].append(word)
print(dict(groups))

# setdefault — те саме без defaultdict
d2 = {}
d2.setdefault("a", []).append(1)
print(d2)
```

## 4. Comprehensions (включення)

Компактний і швидкий спосіб створити колекцію на основі іншої. Читається як «збери X для кожного Y, де умова».

```python
# list comprehension
squares = [n ** 2 for n in range(1, 6)]
print(squares)                                # [1, 4, 9, 16, 25]

# з умовою (фільтрація)
evens = [n for n in range(10) if n % 2 == 0]
print(evens)                                  # [0, 2, 4, 6, 8]

# dict comprehension
prices = {"кава": 45, "чай": 30}
with_vat = {name: round(p * 1.2, 2) for name, p in prices.items()}
print(with_vat)

# set comprehension — унікальні довжини слів
lengths = {len(w) for w in ["a", "bb", "cc", "ddd"]}
print(lengths)                                # {1, 2, 3}
```

## 5. Копіювання: поверхневе vs глибоке ⭐

- `b = a` — **не копія**, а другий ярлик на той самий об'єкт (див. Лекцію 1).
- **Поверхнева копія** (`a.copy()`, `list(a)`, `a[:]`) — новий зовнішній об'єкт, але **вкладені** об'єкти спільні.
- **Глибока копія** (`copy.deepcopy(a)`) — рекурсивно копіює все.

```python
import copy

original = [[1, 2], [3, 4]]

shallow = original.copy()          # поверхнева
shallow[0].append(99)              # міняємо ВКЛАДЕНИЙ список
print("original після shallow:", original)  # [[1, 2, 99], [3, 4]] — теж змінився!

original = [[1, 2], [3, 4]]
deep = copy.deepcopy(original)     # глибока
deep[0].append(99)
print("original після deep:   ", original)  # [[1, 2], [3, 4]] — не змінився
```

## ⚡ Каверзні питання

**Q1. Чому список не може бути ключем словника, а кортеж — може?**
Ключ має бути **хешованим** (незмінним). `list` змінний → його хеш міг би змінитись після вставки, і словник «загубив» би значення. `tuple` незмінний → хешований (якщо всі його елементи хешовані).

**Q2. Яка складність пошуку `x in d` для dict/set і `x in lst` для списку?**
Для `dict`/`set` — у середньому **O(1)** (хеш-таблиця). Для `list` — **O(n)** (лінійний перебір). Тому для частих перевірок належності обирають `set`/`dict`.

```python
# Q3. Що поверне цей код і чому?
d = {}
d[1] = "a"
d[True] = "b"
print(d)   # ?
```

**Q3 (відповідь).** `{1: 'b'}`. Оскільки `hash(1) == hash(True)` і `1 == True`, Python вважає їх **одним ключем** — друге присвоєння перезаписало значення.

```python
# Q4. Чи гарантований порядок у dict?
d = {"z": 1, "a": 2, "m": 3}
print(list(d))   # ['z', 'a', 'm'] — порядок ВСТАВКИ
```

**Q4 (відповідь).** Так, **порядок вставки** гарантований мовою з Python 3.7 (у 3.6 був як деталь реалізації CPython). Це **не** сортування — саме порядок додавання.

```python
# Q5. Чому shallow-копія "протікає"? Доведіть.
a = [[0]]   
b = list(a)          # поверхнева копія
print("a is b:", a is b)                 # False — різні зовнішні обʼєкти
print("a[0] is b[0]:", a[0] is b[0])     # True  — вкладений обʼєкт СПІЛЬНИЙ
b[0].append(1)
print("a:", a)                            # [[0, 1]] — змінився через b
```

**Q5 (відповідь).** Поверхнева копія копіює лише верхній рівень; вкладені об'єкти залишаються спільними (`a[0] is b[0]`). Для повної незалежності — `copy.deepcopy`.

```python
# Q6. Розворот рядка та перевірка на паліндром одним рядком
s = "радар"
print(s == s[::-1])   # True — зріз [::-1] розвертає послідовність
```

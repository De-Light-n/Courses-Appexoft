### Завдання 2 — Type hints
Додайте анотації типів до функцій (аргументи і значення, що повертається).
Підказка: друга функція може повернути `None`, якщо список порожній — згадайте `Optional`.

```python
# ❌ Без анотацій
def repeat(text, times):
    return text * times

def first_or_none(items):
    return items[0] if items else None

print(repeat("ab", 3))
print(first_or_none([10, 20]), first_or_none([]))
```

```python
# ✅ Ваш варіант з type hints:
```

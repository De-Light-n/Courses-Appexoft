## ⚙️ Підготовка A — sqlite3 (для завдань про індекси)

```python
import sqlite3
raw = sqlite3.connect(":memory:")
raw.execute("CREATE TABLE clients (id INTEGER PRIMARY KEY, email TEXT, city TEXT)")
raw.executemany("INSERT INTO clients (id, email, city) VALUES (?,?,?)",
                [(i, f"c{i}@mail.com", ["Київ","Львів","Одеса"][i % 3]) for i in range(50_000)])
raw.commit()
print("Таблиця clients готова:", raw.execute("SELECT COUNT(*) FROM clients").fetchone()[0], "рядків")
```

## ⚙️ Підготовка B — SQLAlchemy (для завдань про CRUD)

```python
from sqlalchemy import create_engine, String, select
from sqlalchemy.orm import DeclarativeBase, Mapped, mapped_column, Session
from sqlalchemy.pool import StaticPool

engine = create_engine("sqlite://", connect_args={"check_same_thread": False}, poolclass=StaticPool)

class Base(DeclarativeBase):
    pass

class Product(Base):
    __tablename__ = "products"
    id:    Mapped[int] = mapped_column(primary_key=True)
    title: Mapped[str] = mapped_column(String(100))
    price: Mapped[float]
    def __repr__(self):
        return f"Product(id={self.id}, title={self.title!r}, price={self.price})"

Base.metadata.create_all(engine)

with Session(engine) as s:
    s.add_all([Product(title="Клавіатура", price=800),
               Product(title="Миша", price=400),
               Product(title="Монітор", price=6000)])
    s.commit()
print("Таблиця products готова")
```

## 1. Що таке ООП

**Об'єктно-орієнтоване програмування (ООП)** — це спосіб організації коду навколо **об'єктів**.
Об'єкт поєднує **дані** (атрибути) та **поведінку** (методи) в одне ціле.

Уявіть **собаку**: у неї є **характеристики** (ім'я, порода, вік) і **дії** (гавкати, їсти).
В ООП ми описуємо це як один об'єкт «собака», а не як окремі розкидані змінні й функції.

**Навіщо це потрібно:**
- **Порядок:** пов'язані дані й дії — разом, а не розкидані по коду.
- **Повторне використання:** один раз описали «шаблон» — створюємо скільки завгодно об'єктів.
- **Масштабованість:** великий проєкт легше підтримувати й розширювати.

Порівняймо два підходи на прикладі:

```python
# ❌ Процедурний стиль: дані окремо, функції окремо
dog_name = "Рекс"
dog_age = 3

def bark(name):
    return f"{name} гавкає: Гав!"

print(bark(dog_name))

# ✅ ООП-стиль (далі розберемо детально): дані й поведінка разом в об'єкті
class Dog:
    def __init__(self, name, age):
        self.name = name
        self.age = age
    def bark(self):
        return f"{self.name} гавкає: Гав!"

rex = Dog("Рекс", 3)
print(rex.bark())
```

## 2. Класи та об'єкти

- **Клас** — це **шаблон** (креслення), який описує, які дані й дії матиме об'єкт.
- **Об'єкт (екземпляр)** — конкретна річ, створена за цим шаблоном.

Аналогія: **клас** — це креслення будинку, а **об'єкти** — реальні будинки, збудовані за ним.
Клас один, а об'єктів за ним — скільки завгодно.

Клас оголошують словом `class`. Спеціальний метод **`__init__`** — це **конструктор**:
він виконується при створенні об'єкта й задає початкові дані. Перший параметр методів — завжди
**`self`** (посилання на сам об'єкт).

```python
class Dog:
    def __init__(self, name, age):   # конструктор: викликається при створенні об'єкта
        self.name = name             # self.name — атрибут ЦЬОГО об'єкта
        self.age = age

# Створюємо ОБ'ЄКТИ (екземпляри) класу:
dog1 = Dog("Рекс", 3)
dog2 = Dog("Бакс", 5)

print(dog1.name, "-", dog1.age, "р.")   # Рекс - 3 р.
print(dog2.name, "-", dog2.age, "р.")   # Бакс - 5 р.
print(type(dog1))                        # <class '__main__.Dog'>
```

> 🧠 Про `self`: коли ви пишете `dog1.name`, Python сам передає `dog1` як `self`.
> Тому в методах ми звертаємось до даних об'єкта через `self.щось`. Це не магія — просто
> «поточний об'єкт».

## 3. Атрибути та методи

- **Атрибут** — це **дані** об'єкта (`self.name`, `self.age`).
- **Метод** — це **функція всередині класу**, тобто **поведінка** об'єкта.

Методи оголошують як звичайні функції, але з `self` першим параметром.

```python
class Dog:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def bark(self):                       # метод
        return f"{self.name} гавкає: Гав!"

    def birthday(self):                   # метод, що ЗМІНЮЄ атрибут
        self.age += 1
        return f"{self.name} тепер {self.age} р."

rex = Dog("Рекс", 3)
print(rex.bark())        # Рекс гавкає: Гав!
print(rex.birthday())    # Рекс тепер 4 р.
print(rex.age)           # 4 — атрибут змінився
```

### Атрибути класу vs атрибути об'єкта

- **Атрибут об'єкта** (`self.name`) — свій у **кожного** об'єкта.
- **Атрибут класу** — **спільний** для всіх об'єктів (оголошується прямо в тілі класу).

```python
class Dog:
    species = "Собака"          # атрибут КЛАСУ — спільний для всіх собак

    def __init__(self, name):
        self.name = name        # атрибут ОБ'ЄКТА — свій у кожного

a = Dog("Рекс")
b = Dog("Бакс")

print(a.name, b.name)       # Рекс Бакс — різні
print(a.species, b.species) # Собака Собака — спільний
print(Dog.species)          # можна звертатись і через клас
```

> 🎯 **Питання: де «живуть» атрибути об'єкта?** У кожного об'єкта є службовий словник `__dict__`,
> у якому лежать саме **його** атрибути. Коли ви пишете `obj.x`, Python спочатку шукає `x` у
> `obj.__dict__`, а якщо не знаходить — дивиться в **клас** (і далі по ланцюжку наслідування).
> Тому атрибут класу (`species`) є доступним через об'єкт, хоча в самому об'єкті його немає.

```python
a = Dog("Рекс")
print(a.__dict__)                 # {'name': 'Рекс'} — тут лише атрибути ОБ'ЄКТА
print("species" in a.__dict__)    # False — у самому об'єкті species немає...
print(a.species)                  # ...але доступ працює — Python бере його з КЛАСУ
```

## 4. `@classmethod` та `@staticmethod`

Крім звичайних методів (що працюють з об'єктом через `self`), є ще два види:

- **`@classmethod`** — отримує не об'єкт, а **сам клас** (перший параметр `cls`). Часто
  використовують як **альтернативний конструктор** (інший спосіб створити об'єкт).
- **`@staticmethod`** — не отримує ні `self`, ні `cls`. Це просто **функція-помічник**,
  логічно пов'язана з класом, але яка не потребує даних об'єкта.

```python
class Dog:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    @classmethod
    def puppy(cls, name):
        # Альтернативний конструктор: цуценя завжди має вік 0.
        # cls — це сам клас Dog, тож cls(...) створює новий об'єкт.
        return cls(name, 0)

    @staticmethod
    def is_adult(age):
        # Помічник: не потребує конкретного об'єкта, лише вхідне число.
        return age >= 2

d1 = Dog("Рекс", 5)             # звичайне створення
d2 = Dog.puppy("Мухтар")        # через classmethod-конструктор

print(d2.name, d2.age)          # Мухтар 0
print("Рекс дорослий?", Dog.is_adult(d1.age))   # True
print("Мухтар дорослий?", Dog.is_adult(d2.age)) # False
```

> 🎯 **Питання: навіщо `@classmethod`, якщо є `__init__`?** Щоб дати **кілька зручних способів
> створити об'єкт** (альтернативні конструктори). `__init__` один, а «фабричних» методів може
> бути багато: `User.from_dict(data)`, `Date.from_string("2026-07-10")`, `Dog.puppy("Мухтар")`.
> Усередині вони кличуть `cls(...)`, тож працюють і для нащадків.
>
> 🎯 **`@staticmethod` vs звичайна функція поза класом?** Різниці в роботі майже немає —
> статичний метод кладуть у клас лише щоб **згрупувати** логічно пов'язану логіку поруч із класом.
> Він не спирається ні на `self`, ні на `cls`.

## 5. Наслідування (inheritance)

**Наслідування** дозволяє створити новий клас на основі наявного: нащадок **отримує** всі
атрибути й методи батька й може додати свої або **перевизначити** наявні.

Це зв'язок **«є різновидом»** (*is-a*): собака **є** твариною, кіт **є** твариною.

```python
# Базовий (батьківський) клас:
class Animal:
    def __init__(self, name):
        self.name = name

    def eat(self):
        return f"{self.name} їсть"

    def speak(self):
        return "..."

# Нащадки: успадковують __init__ та eat(), додають/змінюють свою поведінку
class Dog(Animal):
    def speak(self):                 # перевизначаємо метод батька
        return f"{self.name}: Гав!"

class Cat(Animal):
    def speak(self):
        return f"{self.name}: Няв!"

rex = Dog("Рекс")
murka = Cat("Мурка")

print(rex.eat())       # успадковано від Animal -> Рекс їсть
print(rex.speak())     # власна поведінка -> Рекс: Гав!
print(murka.speak())   # Мурка: Няв!
```

### `super()` — звернутися до батьківського класу

Коли нащадок хоче **доповнити** (а не повністю замінити) поведінку батька, він викликає
батьківський метод через **`super()`**. Найчастіше — у конструкторі `__init__`:

```python
class Animal:
    def __init__(self, name):
        self.name = name

class Dog(Animal):
    def __init__(self, name, breed):
        super().__init__(name)       # викликаємо __init__ батька (задасть self.name)
        self.breed = breed           # і додаємо свій атрибут

rex = Dog("Рекс", "лабрадор")
print(rex.name, "-", rex.breed)      # Рекс - лабрадор
```

> 🎯 **Питання: що буде, якщо в нащадку забути `super().__init__()`?** Батьківський `__init__`
> **не виконається**, тож атрибути, які він задає, просто не з'являться — і ви отримаєте
> `AttributeError` при першому ж зверненні. Саме тому в перевизначеному `__init__` майже завжди
> кличуть `super()`.

```python
class Animal:
    def __init__(self, name):
        self.name = name

class DogBroken(Animal):
    def __init__(self, name, breed):
        # ❌ забули super().__init__(name) -> self.name не буде створено
        self.breed = breed

d = DogBroken("Рекс", "лабрадор")
print("breed:", d.breed)          # лабрадор — цей атрибут задано
try:
    print(d.name)                 # а name немає -> помилка
except AttributeError as e:
    print("Помилка:", e)
```

## 6. Інкапсуляція (encapsulation)

**Інкапсуляція** — це «приховування» внутрішніх даних об'єкта й надання доступу до них лише
через методи. Мета — **захистити дані** від некоректних змін ззовні.

У Python рівні доступу — це **домовленості** (не жорсткі заборони):
- `name` — **публічний** (звичайний доступ).
- `_name` — **захищений** (один `_`): «не чіпайте ззовні, це внутрішнє».
- `__name` — **приватний** (два `_`): Python «ускладнює» доступ ззовні (name mangling).

```python
class BankAccount:
    def __init__(self, owner, balance=0):
        self.owner = owner
        self.__balance = balance          # приватний: не можна псувати напряму ззовні

    def deposit(self, amount):
        if amount <= 0:
            raise ValueError("Сума має бути додатною")
        self.__balance += amount

    def withdraw(self, amount):
        if amount > self.__balance:
            raise ValueError("Недостатньо коштів")
        self.__balance -= amount

    def get_balance(self):                # доступ до балансу — тільки через метод
        return self.__balance

acc = BankAccount("Іван", 100) 
acc.deposit(50)
acc.withdraw(30)
print("Баланс:", acc.get_balance())       # 120
```

```python
# Чому це корисно: некоректну операцію об'єкт НЕ дозволить
acc = BankAccount("Олена", 100)
try:
    acc.withdraw(1000)                     # більше, ніж є
except ValueError as e:
    print("Захист спрацював:", e)

# Прямий доступ до __balance ззовні "ускладнено" (name mangling):
try:
    print(acc.__balance)
except AttributeError as e:
    print("Немає прямого доступу:", e)
```

> 🎯 **Питання: `__balance` справді приватний? Чи можна дістатися приватного атрибута ззовні?**
> **Можна.** У Python **немає справжньої приватності**. `__balance` — це лише угода + механізм
> **name mangling**: усередині класу Python автоматично перейменовує `__balance` на
> `_BankAccount__balance`. Знаючи це «справжнє» ім'я, атрибут можна і прочитати, і навіть
> **перезаписати** ззовні, обійшовши перевірки методів. Тобто `__` не *забороняє* доступ, а
> лише **ускладнює випадковий** доступ і захищає від колізій імен при наслідуванні.
> Порівняйте з Java/C++, де `private` контролює компілятор — у Python усе тримається на
> дисципліні розробника (принцип «ми тут усі дорослі»).

```python
# Доводимо, що приватність — умовна:
acc = BankAccount("Іван", 100)

print(acc._BankAccount__balance)     # 100 — доступ є, треба лише знати "справжнє" ім'я
acc._BankAccount__balance = 999999   # можна навіть перезаписати в обхід deposit/withdraw
print(acc.get_balance())             # 999999 — захист методів обійдено

print(acc.__dict__)                  # видно реальне ім'я: {'owner': 'Іван', '_BankAccount__balance': 999999}
```

## 7. Поліморфізм (polymorphism)

**Поліморфізм** («багато форм») — це коли **різні** об'єкти реагують на **той самий** виклик
**по-своєму**. Ми викликаємо `speak()`, і кожна тварина відповідає по-своєму — не перевіряючи
вручну, хто це.

```python
class Dog:
    def speak(self):
        return "Гав!"

class Cat:
    def speak(self):
        return "Няв!"

class Duck:
    def speak(self):
        return "Кря!"

# Той самий код працює для БУДЬ-ЯКОГО об'єкта, що має метод speak():
animals = [Dog(), Cat(), Duck()]
for animal in animals:
    print(animal)
```

> 🔎 **Важливо знати — «качина типізація» (duck typing).** У Python не важливо, якого саме
> класу об'єкт — важливо, щоб у нього був потрібний метод. «Якщо крякає як качка — то й качка».
> Тому поліморфізм у Python працює навіть без спільного батьківського класу.

## 8. Абстракція (abstraction)

**Абстракція** — це виділення **головного** й приховування деталей. На практиці часто означає
створення **абстрактного класу** — «контракту», який описує, **які методи** мають бути, але
не реалізує їх. Створити об'єкт такого класу не можна — лише його нащадків.

Для цього є модуль **`abc`** (Abstract Base Class) і декоратор **`@abstractmethod`**.

```python
from abc import ABC, abstractmethod

class Shape(ABC):                     # абстрактний базовий клас
    @abstractmethod
    def area(self):                   # "контракт": кожна фігура МУСИТЬ мати area()
        ...

class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius
    def area(self):
        return 3.14159 * self.radius ** 2

class Rectangle(Shape):
    def __init__(self, w, h):
        self.w, self.h = w, h
    def area(self):
        return self.w * self.h

for shape in [Circle(5), Rectangle(3, 4)]:
    print(type(shape).__name__, "-> площа:", shape.area())
```

```python
# Абстрактний клас не можна створити напряму — Python не дозволить:
from abc import ABC, abstractmethod

class Shape(ABC):
    @abstractmethod
    def area(self): ...

try:
    s = Shape()
except TypeError as e:
    print("Не можна створити абстрактний клас:", e)
```

> 🎯 **Питання: у Python є інтерфейси?** Окремого ключового слова `interface` немає. Роль
> інтерфейсу виконує **абстрактний клас**, у якому **всі** методи абстрактні (`@abstractmethod`) —
> це чистий «контракт» без реалізації. Нащадок **зобов'язаний** реалізувати всі абстрактні
> методи, інакше його теж не можна створити. Другий, суто пайтонівський спосіб «контракту» —
> **duck typing** (розділ 7): об'єкту достатньо мати потрібні методи, навіть без спільного
> базового класу.

## 9. 🔎 Важливо знати: MRO та `super()`

Клас може успадковувати **від кількох** батьків одночасно (множинне наслідування). Тоді постає
питання: у якому **порядку** Python шукатиме метод? Відповідь дає **MRO (Method Resolution Order)** —
порядок розв'язання методів. Python обчислює його за алгоритмом **C3-лінеаризації**.

Побачити порядок можна через `.__mro__` або `.mro()`.

```python
class A:
    def whoami(self):
        return "A"

class B(A):
    def whoami(self):
        return "B"

class C(A):
    def whoami(self):
        return "C"

class D(B, C):        # множинне наслідування
    pass

d = D()
print("whoami:", d.whoami())     # 'B' — за порядком MRO

# Порядок пошуку методів (MRO):
for cls in D.__mro__:
    print(" ", cls.__name__)     # D -> B -> C -> A -> object
```

**Як це читати:** Python шукає метод у `D`, потім `B`, потім `C`, потім `A`, і нарешті
`object` (базовий клас усіх класів). Перший знайдений — виграє.

`super()` теж працює **за MRO**: він викликає «наступний по порядку» клас, а не обов'язково
прямого батька. Саме тому `super()` коректно працює навіть у складному множинному наслідуванні.

## 10. 🔎 Важливо знати: магічні (dunder) методи

**Магічні методи** (вони ж *dunder* — від *double underscore*, `__ось_так__`) — це методи зі
спеціальними іменами, які Python викликає **автоматично** у відповідь на певну дію з об'єктом.
Ви їх майже ніколи не викликаєте руками — за вас це робить сам синтаксис мови:

| Ви пишете | Python викликає |
|-----------|-----------------|
| `obj = MyClass(...)` | `__init__` |
| `print(obj)` / `str(obj)` | `__str__` |
| `repr(obj)`, вивід у консолі, в колекціях | `__repr__` |
| `len(obj)` | `__len__` |
| `obj1 == obj2` | `__eq__` |
| `obj1 < obj2`, `sorted(...)` | `__lt__` |
| `obj[key]` | `__getitem__` |
| `x in obj` | `__contains__` |
| `obj(...)` | `__call__` |
| `for x in obj` | `__iter__` / `__next__` |
| `with obj:` | `__enter__` / `__exit__` |
| `obj1 + obj2` | `__add__` |

Саме завдяки dunder-методам **ваші класи поводяться як «рідні» типи Python**. Розберемо
найпопулярніші (їх найчастіше й питають на співбесіді).

### `__str__` та `__repr__` — як об'єкт виглядає у виводі

```python
# Без __str__/__repr__ об'єкт друкується непоказово:
class PointBad:
    def __init__(self, x, y):
        self.x = x
        self.y = y

print(PointBad(1, 2))   # <__main__.PointBad object at 0x...> — марно для читання

# З __str__ (для людини) та __repr__ (для розробника/дебагу):
class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __str__(self):                    # print(obj) / str(obj)
        return f"({self.x}, {self.y})"

    def __repr__(self):                   # repr(obj), консоль, вивід у колекціях
        return f"Point(x={self.x}, y={self.y})"

p = Point(1, 2)
print(p)          # (1, 2)            -> __str__
print(repr(p))    # Point(x=1, y=2)   -> __repr__
print([p])        # [Point(x=1, y=2)] -> у списку працює __repr__
```

> 🎯 **Часте питання: `__str__` vs `__repr__`?**
> `__str__` — «людяне», читабельне подання (для користувача, `print`).
> `__repr__` — однозначне, технічне (для дебагу; його ж показує консоль і колекції).
> Хороший `__repr__` виглядає так, щоб по ньому було видно, **як відтворити об'єкт**.
> Якщо визначаєте лише один — робіть `__repr__` (він працює і як запасний `__str__`).

### `__eq__` та `__lt__` — рівність і порівняння за змістом

```python
# За замовчуванням == порівнює АДРЕСИ в пам'яті (як is). __eq__ вмикає порівняння за ЗМІСТОМ:
class Money:
    def __init__(self, amount, currency="UAH"):
        self.amount = amount
        self.currency = currency

    def __eq__(self, other):
        return self.amount == other.amount and self.currency == other.currency

    def __lt__(self, other):              # дає < (а sorted() використає його для сортування)
        return self.amount < other.amount

    def __repr__(self):
        return f"Money({self.amount}, {self.currency!r})"

print(Money(100) == Money(100))           # True  -> __eq__ (без нього було б False)
print(Money(100) < Money(150))            # True  -> __lt__
print(sorted([Money(30), Money(10), Money(20)]))   # sort спирається на __lt__
```

> 🎯 **Пастка: визначив `__eq__` — подумай про `__hash__`.** Коли ви додаєте власний `__eq__`,
> Python **прибирає** стандартний `__hash__`, і об'єкт стає *нехешованим* — його вже не можна
> покласти в `set` чи використати ключем `dict`. Якщо об'єкт має бути хешованим — додайте
> `__hash__` (зазвичай `return hash((поле1, поле2))`). Хешувати можна лише **незмінні** поля.

```python
# Демонстрація: свій __eq__ без __hash__ -> об'єкт не можна класти в set:
class Money:
    def __init__(self, amount):
        self.amount = amount
    def __eq__(self, other):
        return self.amount == other.amount
    # __hash__ не визначено

try:
    {Money(10), Money(20)}                # set вимагає хешованих елементів
except TypeError as e:
    print("Помилка:", e)                  # unhashable type: 'Money'

# Виправлення — додати __hash__:
class MoneyHashable:
    def __init__(self, amount):
        self.amount = amount
    def __eq__(self, other):
        return self.amount == other.amount
    def __hash__(self):
        return hash(self.amount)

# Тепер дублікат за значенням прибирається коректно:
print(len({MoneyHashable(10), MoneyHashable(10), MoneyHashable(20)}))   # 2
```

### Протокол контейнера: `__len__`, `__getitem__`, `__contains__`, `__add__`

```python
# Робимо власний клас, який поводиться як колекція:
class Playlist:
    def __init__(self, songs):
        self.songs = list(songs)

    def __len__(self):                 # len(pl)
        return len(self.songs)

    def __getitem__(self, i):          # pl[0]; а ще робить об'єкт ітерованим у for
        return self.songs[i]

    def __contains__(self, song):      # song in pl
        return song in self.songs

    def __add__(self, other):          # pl1 + pl2
        return Playlist(self.songs + other.songs)

pl = Playlist(["a", "b", "c"])
print(len(pl))            # 3            -> __len__
print(pl[0])             # a            -> __getitem__
print("b" in pl)          # True         -> __contains__
for song in pl:          # __getitem__ дає ітерацію навіть без окремого __iter__
    print(song, end=" ")
print()
print(len(pl + Playlist(["d"])))        # 4  -> __add__
```

### `__call__` — об'єкт, який можна викликати як функцію

```python
class Multiplier:
    def __init__(self, factor):
        self.factor = factor
    def __call__(self, x):          # тепер екземпляр можна викликати: obj(x)
        return x * self.factor

double = Multiplier(2)
print(double(10))        # 20 — виглядає як функція, але це об'єкт зі станом (factor)
print(callable(double))  # True
```

> `__iter__`/`__next__` (власний ітератор) і `__enter__`/`__exit__` (менеджер контексту `with`)
> ви вже бачили в [Модулі 1, Лекція 4](../modul-1-vstup-do-python/lektsiya-4-funktsiyi-oblast-vydymosti-legb.ipynb).
> Це теж магічні методи — тепер видно, звідки береться їхня «магія».

### 🚀 Middle+: `__slots__` і `__new__`

- **`__slots__`** — фіксує дозволений набір атрибутів і **прибирає `__dict__`** об'єкта.
  Наслідок: менше пам'яті і трохи швидший доступ до атрибутів (важливо, коли об'єктів мільйони).
  Junior-у досить знати, що це **оптимізація пам'яті**; деталі — для Middle+.
- **`__new__`** — власне **створює** об'єкт і викликається **до** `__init__` (той лише наповнює
  вже створений). У 99% випадків потрібен тільки `__init__`. На співбесіді досить знати різницю:
  `__new__` **створює**, `__init__` **ініціалізує**.

```python
class WithoutSlots:
    def __init__(self, x):
        self.x = x

class WithSlots:
    __slots__ = ("x",)              # дозволені лише перелічені атрибути
    def __init__(self, x):
        self.x = x

a = WithoutSlots(1)
a.y = 99                            # можна доліпити будь-який атрибут -> є __dict__
print("__dict__:", a.__dict__)

b = WithSlots(1)
try:
    b.y = 99                        # заборонено -> економія пам'яті
except AttributeError as e:
    print("Помилка:", e)
```

## 11. 🎯 Що спитають на співбесіді про класи

Концентрат найчастіших питань. Відповіді короткі — саме такі й очікують почути. Деталі кожної
теми — у розділах вище; тут — «як відповісти за 10 секунд» + дві класичні **пастки**, на яких
валяться навіть ті, хто «знає ООП».

| Питання | Коротка відповідь |
|---------|-------------------|
| Клас vs об'єкт? | Клас — шаблон/креслення; об'єкт — конкретний екземпляр за цим кресленням |
| Що таке `self`? | Сам об'єкт; Python передає його автоматично першим аргументом методу |
| `__init__` створює об'єкт? | Ні, лише **ініціалізує**; створює об'єкт `__new__` |
| `__str__` vs `__repr__`? | Для людини (`print`) vs для розробника/дебагу (`repr`, колекції) |
| `==` vs `is`? | Рівність за **значенням** (`__eq__`) vs **той самий** об'єкт у пам'яті |
| Визначив `__eq__` — а `__hash__`? | Об'єкт стає нехешованим; додай `__hash__`, якщо треба в `set`/`dict` |
| `_name` vs `__name`? | Угода «не чіпай» ззовні vs **name mangling** (`__name` → `_Клас__name`) |
| `@classmethod` vs `@staticmethod`? | `cls`, часто альт-конструктор vs без `self`/`cls`, просто помічник |
| Наслідування vs композиція? | «**є** різновидом» vs «**має** у складі»; частіше обирають композицію |
| Абстракція / інтерфейс у Python? | `abc.ABC` + `@abstractmethod` (контракт, який зобов'язані реалізувати) |
| Що таке MRO? | Порядок пошуку методів при множинному наслідуванні (C3), `.__mro__` |
| 4 стовпи ООП? | Наслідування, інкапсуляція, поліморфізм, абстракція |
| Що таке duck typing? | Важливий не клас, а наявність потрібного методу («крякає як качка — качка») |

### ⚠️ Пастка 1 — mutable-атрибут **класу** спільний для всіх об'єктів

Класика співбесіди. Якщо змінюваний (`list`, `dict`, `set`) атрибут оголосити **на рівні класу**,
він **один на всіх** — зміна через один об'єкт видно в усіх.

```python
class Team:
    members = []                     # ❌ атрибут КЛАСУ — один список на всі об'єкти
    def __init__(self, name):
        self.name = name

a = Team("A")
b = Team("B")
a.members.append("Іван")             # думали, що додали лише в a...
print("b бачить:", b.members)        # ['Іван'] — спільний список!

# ПРАВИЛЬНО: змінюваний стан створюємо в __init__ (атрибут ОБ'ЄКТА)
class TeamOk:
    def __init__(self, name):
        self.name = name
        self.members = []            # свій список у кожного об'єкта

x, y = TeamOk("A"), TeamOk("B")
x.members.append("Іван")
print("y бачить:", y.members)        # [] — об'єкти незалежні
```

### ⚠️ Пастка 2 — змінюваний аргумент за замовчуванням

Значення за замовчуванням обчислюється **один раз** при оголошенні методу, а не на кожен виклик.
Тому спільний `[]` «пам'ятає» зміни між викликами.

```python
class Cart:
    def __init__(self, items=[]):        # ❌ один список на всі виклики без items
        self.items = items

c1 = Cart()
c1.items.append("яблуко")
print("новий Cart:", Cart().items)       # ['яблуко'] — сюрприз!

# ПРАВИЛЬНО: default = None, а список створюємо всередині
class CartOk:
    def __init__(self, items=None):
        self.items = items if items is not None else []

d1 = CartOk()
d1.items.append("яблуко")
print("новий CartOk:", CartOk().items)   # [] — чисто
```

# ✅ Підсумок уроку

- **ООП** організовує код навколо **об'єктів** = дані (атрибути) + поведінка (методи).
- **Клас** — шаблон, **об'єкт** — екземпляр; `__init__` — конструктор; `self` — сам об'єкт.
- **Атрибути класу** спільні; **атрибути об'єкта** — свої в кожного.
- **`@classmethod`** (`cls`) — часто альтернативний конструктор; **`@staticmethod`** — помічник.
- **4 стовпи ООП:**
  - **Наслідування** — «є різновидом»; `super()` кличе батька.
  - **Інкапсуляція** — приховування даних (`_`, `__`), доступ через методи.
  - **Поліморфізм** — однаковий виклик, різна поведінка; duck typing.
  - **Абстракція** — контракт через `abc` / `@abstractmethod`.
- **MRO** — порядок пошуку методів при множинному наслідуванні (`.__mro__`).
- **Магічні (dunder) методи** роблять клас «рідним» типом: `__str__`/`__repr__`, `__eq__`
  (+`__hash__`!), `__lt__`, `__len__`, `__getitem__`, `__contains__`, `__call__`, `__add__`.
- **Пастки співбесіди:** mutable-атрибут **класу** спільний для всіх; змінюваний аргумент
  за замовчуванням (`items=[]`) «пам'ятає» зміни — роби `=None` і створюй усередині.

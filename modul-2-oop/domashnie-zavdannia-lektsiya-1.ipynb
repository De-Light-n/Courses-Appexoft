### Завдання 3 — @classmethod

Додайте до `Car` classmethod `electric(brand)`, який створює авто з `year=2024`. _(Розділ 4.)_

```python
# Ваш код тут
class Car:
    def __init__(self, brand, year):
        self.brand = brand
        self.year = year
    
    @classmethod
    def electric(cls, brand):
        return cls(brand, 2024)
    
    def info(self):
        return f"Інфо про авто: {self.brand} {self.year}"
    
c1 = Car("BMW", 2020)
c2 = Car.electric("Audi")

print(c1.brand, c2.year)
print(c1.info())
print(c2.info())


# Я трохи експерементував щоб більше зрозуміти
```

### Завдання 4 — @staticmethod

Додайте staticmethod `is_vintage(year)`, що повертає `True`, якщо `year < 2000`. _(Розділ 4.)_

```python
# Ваш код тут
class Car:
    def __init__(self, brand, year):
        self.brand = brand
        self.year = year
    
    @staticmethod
    def is_vintage(year):
        return year < 2000
    
c1 = Car("BMW", 2020)
c2 = Car("Audi", 1999)

print(Car.is_vintage(c1.year))
print(Car.is_vintage(c2.year))
```

### Завдання 5 — Наслідування + super()

Створіть базовий клас `Vehicle` з `__init__(self, brand)`. Створіть нащадка `Truck(Vehicle)`, який у `__init__` через `super()` задає `brand` і додає свій атрибут `capacity`. _(Розділ 5.)_

```python
# Ваш код тут
class Vehicle:
    def __init__(self, brand):
        self.brand = brand
    
class Truck(Vehicle):
    def __init__(self, brand, capacite):
        super().__init__(brand)
        self.capacity = capacite
        
bmw = Truck("BMW", 4)
print("Авто має стільки то місця: " + bmw.brand, bmw.capacity)
    
    
```

### Завдання 6 — Інкапсуляція

Створіть клас `BankAccount` із приватним `__balance`, методами `deposit`, `withdraw` (з перевіркою) та `get_balance`. Продемонструйте роботу. _(Розділ 6.)_

```python
# Ваш код тут
class BankAccount:
    def __init__(self, name, balance = 0):
        self.name = name
        self.__balance = balance
        
    def deposit(self, amount):
        if amount <= 0:
            raise ValueError("Сума має бути додатною")
        self.__balance += amount
        
    def withdraw(self, amount):
        if amount > self.__balance:
            raise ValueError("Недостатньо коштів")
        self.__balance -= amount
        
    def get_balance(self):
        return self.__balance
    
acc = BankAccount("Bohdan", 1000) 
acc.deposit(500)
acc.withdraw(1000)
print("Баланс:", acc.get_balance()) 
```

### Завдання 8 — Абстракція

Створіть абстрактний клас `Shape` (через `abc`) з абстрактним методом `area()`. Реалізуйте `Square` і `Triangle`. Порахуйте площі. _(Розділ 8.)_

```python
# Ваш код тут
from abc import ABC, abstractmethod

class Shape(ABC):
    @abstractmethod
    def area(self):
        ...

class Square(Shape):
    def __init__(self, side):
        self.side = side
    def area(self):
        return self.side ** 2
    
class Triangle(Shape):
    def __init__(self, w, h):
        self.w, self.h = w, h
    def area(self):
        return self.w * self.h / 2
    
for shape in [Square(5), Triangle(3, 4)]:
    print(type(shape).__name__, "-> площа:", shape.area())
    
          
```

### Завдання 9 (⭐) — MRO

Створіть класи `A`, `B(A)`, `C(A)`, `D(B, C)`. Виведіть `D.__mro__` і поясніть коментарем порядок. _(Розділ 9.)_

```python
# Ваш код тут
class A:
    def whoami(self):
        return "A"
    
class B(A):
    def whoami(self):
        return "B"
    
class C(A):
    def whoami(self):
        return "C"
    
class D(B,C):
    pass

d = D()
print("whoami", d.whoami())

for cls in D.__mro__:
    print(" ", cls.__name__)
    
# MRO: D -> B -> C -> A -> object
# Python шукає метод whoami() саме в такому порядку.
# Знаходить його в B, тому до C і A вже не доходить.
```

### Завдання 1 — Клас Car

Створіть клас `Car` з атрибутами `brand`, `year` (задаються в `__init__`) та методом `info()`, що повертає рядок `"Toyota, 2020"`. Створіть 2 об'єкти й виведіть їхнє info. _(Розділи 2–3.)_

```python
class Car:
    def __init__(self, brand, year):
        self.brand = brand
        self.year = year

    def info(self):
        return f"{self.brand}, {self.year}"
    
car1 = Car("Toyota", 2020)
car2 = Car('Smart OneTwo', 2000)
print(car1.info())
print(car2.info())

```

### Завдання 3 — @classmethod

Додайте до `Car` classmethod `electric(brand)`, який створює авто з `year=2024`. _(Розділ 4.)_

```python
class Car:
    wheels = 4

    def __init__(self, brand, year):
        self.brand = brand
        self.year = year

    @classmethod
    def electric(cls, brand):
        return cls(brand, 2024)
    
    def info(self):
        return f"{self.brand}, {self.year}"

tesla = Car.electric("Tesla")
print(tesla.info())
```

### Завдання 4 — @staticmethod

Додайте staticmethod `is_vintage(year)`, що повертає `True`, якщо `year < 2000`. _(Розділ 4.)_

```python
class Car:
    wheels = 4

    def __init__(self, brand, year):
        self.brand = brand
        self.year = year

    @staticmethod
    def is_vintage(year):
        return year < 2000
    
    def info(self):
        return f"{self.brand}, {self.year}"

print(Car.is_vintage(1994))
print(Car.is_vintage(2001))
```

### Завдання 5 — Наслідування + super()

Створіть базовий клас `Vehicle` з `__init__(self, brand)`. Створіть нащадка `Truck(Vehicle)`, який у `__init__` через `super()` задає `brand` і додає свій атрибут `capacity`. _(Розділ 5.)_

```python
class Vehicle:
    def __init__(self, brand):
        self.brand = brand

class Truck(Vehicle):
    def __init__(self, brand, capacity):
        super().__init__(brand)
        self.capacity = capacity


truck = Truck("Volvo", 30)

print(truck.brand)
print(truck.capacity)
```

### Завдання 6 — Інкапсуляція

Створіть клас `BankAccount` із приватним `__balance`, методами `deposit`, `withdraw` (з перевіркою) та `get_balance`. Продемонструйте роботу. _(Розділ 6.)_

```python
class BankAccount:
    def __init__(self, owner, balance):
        self.owner = owner
        self.__balance = balance

    def deposit(self, amount):
        if amount > 0:
            self.__balance = self.__balance + amount
        else:
            print("Сума має бути більшою за 0")

    def withdraw(self, amount):
        if amount > 0 and amount <= self.__balance:
            self.__balance = self.__balance - amount
        else:
            print("Недостатньо коштів або неправильна сума")

    def get_balance(self):
        return self.__balance
    
account = BankAccount("Anastasia", 1000)
print(account.get_balance())
account.deposit(1300)
account.withdraw(269)
print(account.get_balance())
```

### Завдання 7 — Поліморфізм

Створіть класи `Dog`, `Cat`, `Cow` з методом `sound()`. Складіть список об'єктів і в циклі викличте `sound()` для кожного. _(Розділ 7.)_

```python
class Dog:
    def sound(self):
        return 'woof!'
    
class Cat:
    def sound(self):
        return 'meow!'

class Cow:
    def sound(self):
        return 'moo!'
    
dog = Dog()
cat = Cat()
cow = Cow()

animals = [dog, cat, cow]

for animal in animals:
    print(animal.sound())
```

### Завдання 8 — Абстракція

Створіть абстрактний клас `Shape` (через `abc`) з абстрактним методом `area()`. Реалізуйте `Square` і `Triangle`. Порахуйте площі. _(Розділ 8.)_

```python
from abc import ABC, abstractmethod

class Shape(ABC):
    @abstractmethod
    def area(self):
        pass

class Square(Shape):
    def __init__(self, side):
        self.side = side

    def area(self):
        return self.side ** 2
    
class Triangle(Shape):
    def __init__(self, base, height):
        self.base = base
        self.height = height

    def area(self):
        return (self.base * self.height) / 2
    
square = Square(6)
triangle = Triangle(8, 9)

print(square.area())
print(int(triangle.area()))
```

### Завдання 1 — Клас Car

Створіть клас `Car` з атрибутами `brand`, `year` (задаються в `__init__`) та методом `info()`, що повертає рядок `"Toyota, 2020"`. Створіть 2 об'єкти й виведіть їхнє info. _(Розділи 2–3.)_

```python
class Car:
    def __init__(self, brand, year):
        self.brand = brand
        self.year = year

    def info(self):
        return f"{self.brand}, {self.year}"

car1 = Car("Toyota", 2020)
car2 = Car("Nissan", 2018)

print(car1.info())
print(car2.info())
```

### Завдання 2 — Атрибут класу

Додайте до `Car` атрибут класу `wheels = 4` (спільний). Виведіть його через об'єкт і через клас. _(Розділ 3.)_

```python
class Car:
    wheels = 4

    def __init__(self, brand, year):
        self.brand = brand
        self.year = year

    def info(self):
        return f"{self.brand}, {self.year}"

car = Car("Toyota", 2020)

print(Car.wheels)
print(car.wheels)
```

### Завдання 3 — @classmethod

Додайте до `Car` classmethod `electric(brand)`, який створює авто з `year=2024`. _(Розділ 4.)_

```python
class Car:
    wheels = 4

    def __init__(self, brand, year):
        self.brand = brand
        self.year = year

    def info(self):
        return f"{self.brand}, {self.year}"

    @classmethod
    def electric(cls, brand):
        return cls(brand, 2024)

car = Car.electric("Tesla")
print(car.info())
```

### Завдання 5 — Наслідування + super()

Створіть базовий клас `Vehicle` з `__init__(self, brand)`. Створіть нащадка `Truck(Vehicle)`, який у `__init__` через `super()` задає `brand` і додає свій атрибут `capacity`. _(Розділ 5.)_

```python
class Vehicle:

    def __init__(self, brand):
        self.brand = brand


class Truck(Vehicle):

    def __init__(self, brand, capacity):
        super().__init__(brand)
        self.capacity = capacity


truck = Truck("Volvo", 20)
print(truck.brand, truck.capacity)
```

### Завдання 6 — Інкапсуляція

Створіть клас `BankAccount` із приватним `__balance`, методами `deposit`, `withdraw` (з перевіркою) та `get_balance`. Продемонструйте роботу. _(Розділ 6.)_

```python
class BankAccount:

    def __init__(self, balance=0):
        self.__balance = balance

    def deposit(self, amount):
        self.__balance += amount

    def withdraw(self, amount):
        if amount <= self.__balance:
            self.__balance -= amount

    def get_balance(self):
        return self.__balance


account = BankAccount(100)
account.deposit(50)
account.withdraw(30)
print(account.get_balance())
```

### Завдання 7 — Поліморфізм

Створіть класи `Dog`, `Cat`, `Cow` з методом `sound()`. Складіть список об'єктів і в циклі викличте `sound()` для кожного. _(Розділ 7.)_

```python
class Dog:

    def sound(self):
        return "Woof"


class Cat:

    def sound(self):
        return "Meow"


class Cow:

    def sound(self):
        return "Moo"


animals = [Dog(), Cat(), Cow()]

for animal in animals:
    print(animal.sound())
```

### Завдання 8 — Абстракція

Створіть абстрактний клас `Shape` (через `abc`) з абстрактним методом `area()`. Реалізуйте `Square` і `Triangle`. Порахуйте площі. _(Розділ 8.)_

```python
from abc import ABC, abstractmethod


class Shape(ABC):

    @abstractmethod
    def area(self):
        pass


class Square(Shape):

    def __init__(self, side):
        self.side = side

    def area(self):
        return self.side**2


class Triangle(Shape):

    def __init__(self, base, height):
        self.base = base
        self.height = height

    def area(self):
        return 0.5 * self.base * self.height


square = Square(4)
triangle = Triangle(6, 3)

print(square.area())
print(triangle.area())
```

### Завдання 9 (⭐) — MRO

Створіть класи `A`, `B(A)`, `C(A)`, `D(B, C)`. Виведіть `D.__mro__` і поясніть коментарем порядок. _(Розділ 9.)_

```python
class A:
    pass


class B(A):
    pass


class C(A):
    pass


class D(B, C):
    pass


print(D.__mro__)

# Порядок з низу вверх з ліва на право: D -> B -> C -> A -> obj
# Пайтон спочатку шукає метод у самому класі D.
# D успадковує (B, C), спочатку перевіряється ліва гілка (B).
# Далі пошук переходить до сусіднього батька C,
# і тільки після цього піднімається до їхнього спільного предка A, а в кінці до базового класу object.
```

### Завдання 1 — Клас Car

Створіть клас `Car` з атрибутами `brand`, `year` (задаються в `__init__`) та методом `info()`, що повертає рядок `"Toyota, 2020"`. Створіть 2 об'єкти й виведіть їхнє info. _(Розділи 2–3.)_

```python
class Car:
    def __init__(self, brand, year):
        self.brand = brand
        self.year = year


    def info(self):
        return f"{self.brand}, {self.year}"


car1 = Car("Toyota", 2020)
car2 = Car("Ford", 2015)
print(car1.info())
print(car2.info())
```

### Завдання 3 — @classmethod

Додайте до `Car` classmethod `electric(brand)`, який створює авто з `year=2024`. _(Розділ 4.)_

```python
class Car:
    wheels = 4
    
    def __init__(self, brand, year):
        self.brand = brand
        self.year = year


    def info(self):
        print(f"{self.brand}, {self.year}")
        
    @classmethod
    def electric(cls, brand):
        return cls(brand, year = 2024)
    
    
car2 = Car.electric("Toyota")
car2.info()
```

### Завдання 4 — @staticmethod

Додайте staticmethod `is_vintage(year)`, що повертає `True`, якщо `year < 2000`. _(Розділ 4.)_

```python
class Car:
    wheels = 4
    
    def __init__(self, brand, year):
        self.brand = brand
        self.year = year


    def info(self):
        print(f"{self.brand}, {self.year}")
        
    @classmethod
    def electric(cls, brand):
        return cls(brand, year = 2024)
    
    @staticmethod
    def is_vintage(year):
        return year < 2000
    
    
car1 = Car("Toyota", 1998)
print("Car is vintage:", car1.is_vintage(car1.year))
car2 = Car("Toyota", 2001)
print("Car is vintage:", car2.is_vintage(car2.year))
```

### Завдання 5 — Наслідування + super()

Створіть базовий клас `Vehicle` з `__init__(self, brand)`. Створіть нащадка `Truck(Vehicle)`, який у `__init__` через `super()` задає `brand` і додає свій атрибут `capacity`. _(Розділ 5.)_

```python
class Vehicle:
    def __init__(self, brand):
        self.brand = brand



class Truck(Vehicle):
    def __init__(self, brand, capacity):
        super().__init__(brand)
        self.capacity = capacity
        

truck1 = Truck("Ford", 1000)
print(truck1.brand)
print(truck1.capacity)
```

### Завдання 6 — Інкапсуляція

Створіть клас `BankAccount` із приватним `__balance`, методами `deposit`, `withdraw` (з перевіркою) та `get_balance`. Продемонструйте роботу. _(Розділ 6.)_

```python
class BankAccount:
    def __init__(self, balance):
        self.__balance = balance 

   
    def deposit(self, amount):
        if amount <=0:
            print("Amount must be positive!")
        else:
            self.__balance += amount


    def withdraw(self, amount):
        if amount > self.__balance:
            print("Not enough funds!")
        else:
            self.__balance -= amount


    def get_balance(self):
        return self.__balance
        
        
account = BankAccount(1000)
print(account.get_balance())
account.withdraw(1200)
account.withdraw(400)
account.deposit(-100)
account.deposit(700)
print(account.get_balance())
```

### Завдання 8 — Абстракція

Створіть абстрактний клас `Shape` (через `abc`) з абстрактним методом `area()`. Реалізуйте `Square` і `Triangle`. Порахуйте площі. _(Розділ 8.)_

```python
from abc import ABC, abstractmethod
from math import sqrt

class Shape(ABC):
    @abstractmethod
    def area(self):
       "..."


class Square(Shape):
    def __init__(self, a):
       if a <= 0:
           raise ValueError("Side length must be positive")
       self.a = a


    def area(self):
       return self.a**2


class Triangle(Shape):
    def __init__(self, a, b, c):
        if a <= 0 or b <= 0 or c <= 0:
            raise ValueError("Side length must be positive")
        if a + b < c or a + c < b or b + c < a:
            raise ValueError("Triangle does not exist")
        

        self.a = a
        self.b = b
        self.c = c


    def area(self):
        p = (self.a + self.b + self.c) / 2
        return sqrt(p * (p - self.a) * (p - self.b) * (p - self.c))
    
    
triangle = Triangle(3, 4, 5)
square = Square(8)
print("Triangle area:", triangle.area())
print("Square area:", square.area())
```
